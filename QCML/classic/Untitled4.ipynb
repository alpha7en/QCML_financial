```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, Subset
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
# >>> НОВЫЕ ИМЕНА для CNN-GRU модели <<<
model_save_name = 'cnn_gru_nn_improved.pth'
scaler_save_name = 'scaler_cnn_gru_improved.pth'

target_col = 'IMO_fwd_log_ret_3d' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ

# Параметры последовательности
SEQUENCE_LENGTH = 20 # Можно попробовать немного увеличить окно

# Параметры разделения и обучения
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 3
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 64
learning_rate = 0.001
epochs = 220 # Макс. кол-во эпох

# >>> ПАРАМЕТРЫ CNN-GRU МОДЕЛИ <<<
CNN_OUT_CHANNELS = 32 # Количество фильтров (признаков) после CNN
CNN_KERNEL_SIZE = 3   # Размер окна свертки (например, 3 или 5)
GRU_HIDDEN_SIZE = 64 # Размер скрытого состояния GRU
GRU_NUM_LAYERS = 1   # Попробуем 1 слой GRU после CNN
MLP_HIDDEN_SIZE = 32 # Размер скрытого слоя в MLP после GRU
DROPOUT_RATE = 0.3   # Dropout

# Параметры улучшений
EARLY_STOPPING_PATIENCE = 500
SCHEDULER_PATIENCE = 100
SCHEDULER_FACTOR = 0.1
WEIGHT_DECAY = 1e-5

random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled = df_cleaned[target_col].values.reshape(-1, 1)
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей ---
def create_sequences(input_data, target_data, sequence_length):
    sequences, targets = [], []
    for i in range(sequence_length, len(input_data)):
        sequences.append(input_data[i-sequence_length:i])
        targets.append(target_data[i])
    return np.array(sequences), np.array(targets)
X_seq, y_seq = create_sequences(X_scaled, y_unscaled, SEQUENCE_LENGTH)
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ ---
print("-" * 30); print(f"Разделение последовательностей на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ...")
X_temp, X_final_test, y_temp, y_final_test, temp_indices, final_test_indices = train_test_split(
    X_seq, y_seq, original_indices_seq, test_size=FINAL_TEST_SET_SIZE, random_state=random_seed, shuffle=True
)
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)


# --- 5. Определение НОВОЙ МОДЕЛИ (CNN-GRU) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate

        # 1D CNN Слой
        # Вход: (N, C_in, L_in) = (batch_size, input_dim, sequence_length)
        # Выход: (N, C_out, L_out) = (batch_size, cnn_out_channels, sequence_length) (с padding='same')
        # Нам нужно сначала поменять местами sequence_length и input_dim
        self.conv1d = nn.Conv1d(in_channels=input_dim,
                                out_channels=cnn_out_channels,
                                kernel_size=kernel_size,
                                padding='same') # 'same' сохраняет длину последовательности
        self.bn1 = nn.BatchNorm1d(cnn_out_channels) # BatchNorm применяется к каналам
        self.relu1 = nn.ReLU()
        self.dropout_cnn = nn.Dropout(dropout_rate)

        # GRU Слой
        # Вход: (N, L, H_in) = (batch_size, sequence_length, cnn_out_channels)
        # Выход hn: (num_layers, N, H_out) = (gru_num_layers, batch_size, gru_hidden_dim)
        # Нам нужно снова поменять местами измерения после CNN
        self.gru = nn.GRU(input_size=cnn_out_channels, # Входной размер GRU = выход CNN
                          hidden_size=gru_hidden_dim,
                          num_layers=gru_num_layers,
                          batch_first=True, # Ожидаем (batch, seq, feature)
                          dropout=dropout_rate if gru_num_layers > 1 else 0)

        # MLP Голова
        self.mlp = nn.Sequential(
            nn.Linear(gru_hidden_dim, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(mlp_hidden, output_dim)
        )

    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_dim)

        # --- CNN Часть ---
        # Меняем местами измерения для Conv1d: (N, C_in, L_in)
        x_cnn = x.permute(0, 2, 1)
        x_cnn = self.conv1d(x_cnn)
        x_cnn = self.bn1(x_cnn)
        x_cnn = self.relu1(x_cnn)
        x_cnn = self.dropout_cnn(x_cnn)
        # Меняем местами обратно для GRU: (N, L_out, C_out)
        x_gru_input = x_cnn.permute(0, 2, 1)

        # --- GRU Часть ---
        # Инициализация скрытого состояния (не обязательно, GRU сделает это по умолчанию)
        # h0 = torch.zeros(self.gru.num_layers, x.size(0), self.gru.hidden_size).to(x.device)
        # out_gru, hn = self.gru(x_gru_input, h0)
        out_gru, hn = self.gru(x_gru_input) # Используем выход последнего шага

        # --- MLP Часть ---
        # Берем выход GRU с последнего временного шага
        # out_gru shape: (batch_size, sequence_length, gru_hidden_dim)
        last_time_step_out = out_gru[:, -1, :]
        # Или можно взять последнее скрытое состояние: hn[-1] для последнего слоя
        # last_time_step_out = hn[-1] # shape: (batch_size, gru_hidden_dim)

        output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2] # Количество признаков

# --- 6. Кросс-валидация ---
kf = KFold(n_splits=K_FOLDS, shuffle=True, random_state=random_seed)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
for fold, (train_idx, val_idx) in enumerate(kf.split(X_temp, y_temp)):
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    # Подготовка данных
    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]
    X_train_fold_tensor = torch.tensor(X_train_fold, dtype=torch.float32).to(device)
    y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold, dtype=torch.float32).to(device)
    y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)
    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True)
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False)

    # >>> Инициализация CNN-GRU модели <<<
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

    # Обучение фолда
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1} (модель CNN-GRU)...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y)
            loss.backward(); optimizer.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0
        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        with torch.no_grad():
            for batch_X_val, batch_y_val in val_fold_loader:
                outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
        avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
        if (epoch + 1) % 1 == 0: print(f'    Фолд {fold+1}, Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        scheduler.step(avg_val_loss)
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено на эпохе {epoch+1} (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель фолда.")
    model.eval()
    with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
    mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
    print(f"  Метрики валидации фолда {fold+1} (лучшая): MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
    fold_mses.append(mse_fold); fold_r2s.append(r2_fold)

# --- Результаты кросс-валидации ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
print(f"Среднее MSE по фолдам: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})")
print(f"Среднее R² по фолдам: {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
print(f"Среднее количество 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")

# --- 7. Обучение ФИНАЛЬНОЙ модели ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU NN + Улучшения) ---")
# Подготовка данных
X_train_final_seq, X_val_final_seq, y_train_final_seq, y_val_final_seq = train_test_split(X_temp, y_temp, test_size=VAL_SPLIT_FOR_FINAL, random_state=random_seed, shuffle=True)
X_train_final_tensor = torch.tensor(X_train_final_seq, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
X_val_final_tensor = torch.tensor(X_val_final_seq, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor); final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True)
final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False)

# >>> Инициализация CNN-GRU финальной модели <<<
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss()
final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
final_scheduler = ReduceLROnPlateau(final_optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

# Обучение финальной модели
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения (модель CNN-GRU)...")
start_time_final = time.time()
for epoch in range(epochs):
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y)
        loss.backward(); final_optimizer.step()
        train_loss_epoch += loss.item(); batches_processed_train += 1
    avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in final_val_loader:
            outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
            val_loss_epoch += loss_val.item(); batches_processed_val += 1
    avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
    if (epoch + 1) % 1 == 0: print(f'    Финал. обуч., Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')
    final_scheduler.step(avg_val_loss)
    if avg_val_loss < best_val_loss_final: best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка фин. обучения!"); break
end_time_final = time.time(); print(f"Финальное обучение завершено на эпохе {epoch+1} (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели ---
if best_model_state_final:
    try:
        torch.save(best_model_state_final, MODEL_SAVE_PATH); print(f"Лучшая финальная модель ({best_epoch_final}) сохранена: {MODEL_SAVE_PATH}")
        final_model.load_state_dict(best_model_state_final)
    except Exception as e: print(f"Ошибка сохранения/загрузки: {e}")
else: print("Предупреждение: Используется последняя модель.")

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ + Сравнение с SMA(5) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
final_model.eval()
# ВАЖНО: X_final_test УЖЕ содержит ПОСЛЕДОВАТЕЛЬНОСТИ масштабированных данных
X_final_test_tensor_eval = torch.tensor(X_final_test, dtype=torch.float32).to(device)
with torch.no_grad(): y_pred_nn = final_model(X_final_test_tensor_eval).cpu().numpy()
# Получение предсказаний SMA(5) для соответствующих дат КОНЦА последовательностей
y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
else: print("NaN в SMA(5) не обнаружено.")

# Расчет метрик
mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
ssr_model = mse_final_nn * len(y_final_test)
mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0

print("\n--- Результаты оценки (CNN-GRU NN + Улучшения) ---")
print(f"MSE NN:               {mse_final_nn:.4f}")
print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}")
print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")
print("---")
print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}")
print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}") # <<< СРАВНЕНИЕ С SMA(5)

# --- 10. Визуализация ---
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
plt.plot([y_final_test.min(), y_final_test.max()], [y_final_test.min(), y_final_test.max()], 'r--', lw=2, label='Идеал')
plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест, CNN-GRU)\nЦель: {target_col}")
plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten(), 'Predicted_SMA5': y_pred_sma5.flatten()}, index=final_test_indices).sort_index()
n_points_to_plot = min(100, len(results_final_test_df))
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)\nЦель: {target_col}")
plt.xticks(rotation=45), plt.legend(), plt.grid(True); plt.tight_layout(); plt.show()
plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (CNN-GRU NN)"); plt.legend(), plt.grid(True); plt.show()

# --- (Дополнительно) Пример загрузки ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
try:
    loaded_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                               gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                               mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE)
    try:
        loaded_model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
        loaded_model.to(device); loaded_model.eval()
        print(f"Модель CNN-GRU успешно загружена из {MODEL_SAVE_PATH}")
    except FileNotFoundError: print(f"Файл модели {MODEL_SAVE_PATH} не найден.")
    except Exception as e: print(f"Ошибка при загрузке модели: {e}")
except NameError: print("Переменная 'input_dim' не определена.")
try:
    with open(SCALER_SAVE_PATH, 'rb') as f: loaded_scaler = pickle.load(f)
    print(f"Scaler успешно загружен из {SCALER_SAVE_PATH}")
    # Пример использования:
    # last_sequence = ... # ваши последние SEQUENCE_LENGTH дней признаков, shape (sequence_length, input_dim)
    # last_sequence_scaled = loaded_scaler.transform(last_sequence)
    # new_data_tensor = torch.tensor(last_sequence_scaled, dtype=torch.float32).unsqueeze(0).to(device) # Добавляем batch dimension
    # if 'loaded_model' in locals():
    #     with torch.no_grad(): prediction = loaded_model(new_data_tensor)
    #     print(f"Предсказание для следующего шага: {prediction.item()}")
except FileNotFoundError: print(f"Файл scaler'а {SCALER_SAVE_PATH} не найден.")
except Exception as e: print(f"Ошибка при загрузке scaler'а: {e}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, Subset
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
# >>> НОВЫЕ ИМЕНА для CNN-GRU модели <<<
model_save_name = 'cnn_gru_nn_improved.pth'
scaler_save_name = 'scaler_cnn_gru_improved.pth'

target_col = 'IMO_fwd_log_ret_3d' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ

# Параметры последовательности
SEQUENCE_LENGTH = 15 # Можно попробовать немного увеличить окно

# Параметры разделения и обучения
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 5
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 32
learning_rate = 0.001
epochs = 300 # Макс. кол-во эпох

# >>> ПАРАМЕТРЫ CNN-GRU МОДЕЛИ <<<
CNN_OUT_CHANNELS = 32 # Количество фильтров (признаков) после CNN
CNN_KERNEL_SIZE = 3   # Размер окна свертки (например, 3 или 5)
GRU_HIDDEN_SIZE = 64 # Размер скрытого состояния GRU
GRU_NUM_LAYERS = 1   # Попробуем 1 слой GRU после CNN
MLP_HIDDEN_SIZE = 32 # Размер скрытого слоя в MLP после GRU
DROPOUT_RATE = 0.3   # Dropout

# Параметры улучшений
EARLY_STOPPING_PATIENCE = 50
SCHEDULER_PATIENCE = 10
SCHEDULER_FACTOR = 0.1
WEIGHT_DECAY = 1e-5

random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled = df_cleaned[target_col].values.reshape(-1, 1)
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей ---
def create_sequences(input_data, target_data, sequence_length):
    sequences, targets = [], []
    for i in range(sequence_length, len(input_data)):
        sequences.append(input_data[i-sequence_length:i])
        targets.append(target_data[i])
    return np.array(sequences), np.array(targets)
X_seq, y_seq = create_sequences(X_scaled, y_unscaled, SEQUENCE_LENGTH)
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ ---
print("-" * 30); print(f"Разделение последовательностей на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ...")
X_temp, X_final_test, y_temp, y_final_test, temp_indices, final_test_indices = train_test_split(
    X_seq, y_seq, original_indices_seq, test_size=FINAL_TEST_SET_SIZE, random_state=random_seed, shuffle=True
)
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)


# --- 5. Определение НОВОЙ МОДЕЛИ (CNN-GRU) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate

        # 1D CNN Слой
        # Вход: (N, C_in, L_in) = (batch_size, input_dim, sequence_length)
        # Выход: (N, C_out, L_out) = (batch_size, cnn_out_channels, sequence_length) (с padding='same')
        # Нам нужно сначала поменять местами sequence_length и input_dim
        self.conv1d = nn.Conv1d(in_channels=input_dim,
                                out_channels=cnn_out_channels,
                                kernel_size=kernel_size,
                                padding='same') # 'same' сохраняет длину последовательности
        self.bn1 = nn.BatchNorm1d(cnn_out_channels) # BatchNorm применяется к каналам
        self.relu1 = nn.ReLU()
        self.dropout_cnn = nn.Dropout(dropout_rate)

        # GRU Слой
        # Вход: (N, L, H_in) = (batch_size, sequence_length, cnn_out_channels)
        # Выход hn: (num_layers, N, H_out) = (gru_num_layers, batch_size, gru_hidden_dim)
        # Нам нужно снова поменять местами измерения после CNN
        self.gru = nn.GRU(input_size=cnn_out_channels, # Входной размер GRU = выход CNN
                          hidden_size=gru_hidden_dim,
                          num_layers=gru_num_layers,
                          batch_first=True, # Ожидаем (batch, seq, feature)
                          dropout=dropout_rate if gru_num_layers > 1 else 0)

        # MLP Голова
        self.mlp = nn.Sequential(
            nn.Linear(gru_hidden_dim, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(mlp_hidden, output_dim)
        )

    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_dim)

        # --- CNN Часть ---
        # Меняем местами измерения для Conv1d: (N, C_in, L_in)
        x_cnn = x.permute(0, 2, 1)
        x_cnn = self.conv1d(x_cnn)
        x_cnn = self.bn1(x_cnn)
        x_cnn = self.relu1(x_cnn)
        x_cnn = self.dropout_cnn(x_cnn)
        # Меняем местами обратно для GRU: (N, L_out, C_out)
        x_gru_input = x_cnn.permute(0, 2, 1)

        # --- GRU Часть ---
        # Инициализация скрытого состояния (не обязательно, GRU сделает это по умолчанию)
        # h0 = torch.zeros(self.gru.num_layers, x.size(0), self.gru.hidden_size).to(x.device)
        # out_gru, hn = self.gru(x_gru_input, h0)
        out_gru, hn = self.gru(x_gru_input) # Используем выход последнего шага

        # --- MLP Часть ---
        # Берем выход GRU с последнего временного шага
        # out_gru shape: (batch_size, sequence_length, gru_hidden_dim)
        last_time_step_out = out_gru[:, -1, :]
        # Или можно взять последнее скрытое состояние: hn[-1] для последнего слоя
        # last_time_step_out = hn[-1] # shape: (batch_size, gru_hidden_dim)

        output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2] # Количество признаков

# --- 6. Кросс-валидация ---
kf = KFold(n_splits=K_FOLDS, shuffle=True, random_state=random_seed)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
for fold, (train_idx, val_idx) in enumerate(kf.split(X_temp, y_temp)):
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    # Подготовка данных
    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]
    X_train_fold_tensor = torch.tensor(X_train_fold, dtype=torch.float32).to(device)
    y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold, dtype=torch.float32).to(device)
    y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)
    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True)
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False)

    # >>> Инициализация CNN-GRU модели <<<
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

    # Обучение фолда
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1} (модель CNN-GRU)...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y)
            loss.backward(); optimizer.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0
        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        with torch.no_grad():
            for batch_X_val, batch_y_val in val_fold_loader:
                outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
        avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
        if (epoch + 1) % 10 == 0: print(f'    Фолд {fold+1}, Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        scheduler.step(avg_val_loss)
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено на эпохе {epoch+1} (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель фолда.")
    model.eval()
    with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
    mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
    print(f"  Метрики валидации фолда {fold+1} (лучшая): MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
    fold_mses.append(mse_fold); fold_r2s.append(r2_fold)

# --- Результаты кросс-валидации ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
print(f"Среднее MSE по фолдам: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})")
print(f"Среднее R² по фолдам: {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
print(f"Среднее количество 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")

# --- 7. Обучение ФИНАЛЬНОЙ модели ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU NN + Улучшения) ---")
# Подготовка данных
X_train_final_seq, X_val_final_seq, y_train_final_seq, y_val_final_seq = train_test_split(X_temp, y_temp, test_size=VAL_SPLIT_FOR_FINAL, random_state=random_seed, shuffle=True)
X_train_final_tensor = torch.tensor(X_train_final_seq, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
X_val_final_tensor = torch.tensor(X_val_final_seq, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor); final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True)
final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False)

# >>> Инициализация CNN-GRU финальной модели <<<
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss()
final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
final_scheduler = ReduceLROnPlateau(final_optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

# Обучение финальной модели
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения (модель CNN-GRU)...")
start_time_final = time.time()
for epoch in range(epochs):
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y)
        loss.backward(); final_optimizer.step()
        train_loss_epoch += loss.item(); batches_processed_train += 1
    avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in final_val_loader:
            outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
            val_loss_epoch += loss_val.item(); batches_processed_val += 1
    avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
    if (epoch + 1) % 10 == 0: print(f'    Финал. обуч., Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')
    final_scheduler.step(avg_val_loss)
    if avg_val_loss < best_val_loss_final: best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка фин. обучения!"); break
end_time_final = time.time(); print(f"Финальное обучение завершено на эпохе {epoch+1} (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели ---
if best_model_state_final:
    try:
        torch.save(best_model_state_final, MODEL_SAVE_PATH); print(f"Лучшая финальная модель ({best_epoch_final}) сохранена: {MODEL_SAVE_PATH}")
        final_model.load_state_dict(best_model_state_final)
    except Exception as e: print(f"Ошибка сохранения/загрузки: {e}")
else: print("Предупреждение: Используется последняя модель.")

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ + Сравнение с SMA(5) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
final_model.eval()
# ВАЖНО: X_final_test УЖЕ содержит ПОСЛЕДОВАТЕЛЬНОСТИ масштабированных данных
X_final_test_tensor_eval = torch.tensor(X_final_test, dtype=torch.float32).to(device)
with torch.no_grad(): y_pred_nn = final_model(X_final_test_tensor_eval).cpu().numpy()
# Получение предсказаний SMA(5) для соответствующих дат КОНЦА последовательностей
y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
else: print("NaN в SMA(5) не обнаружено.")

# Расчет метрик
mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
ssr_model = mse_final_nn * len(y_final_test)
mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0

print("\n--- Результаты оценки (CNN-GRU NN + Улучшения) ---")
print(f"MSE NN:               {mse_final_nn:.4f}")
print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}")
print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")
print("---")
print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}")
print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}") # <<< СРАВНЕНИЕ С SMA(5)

# --- 10. Визуализация ---
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
plt.plot([y_final_test.min(), y_final_test.max()], [y_final_test.min(), y_final_test.max()], 'r--', lw=2, label='Идеал')
plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест, CNN-GRU)\nЦель: {target_col}")
plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten(), 'Predicted_SMA5': y_pred_sma5.flatten()}, index=final_test_indices).sort_index()
n_points_to_plot = min(100, len(results_final_test_df))
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)\nЦель: {target_col}")
plt.xticks(rotation=45), plt.legend(), plt.grid(True); plt.tight_layout(); plt.show()
plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (CNN-GRU NN)"); plt.legend(), plt.grid(True); plt.show()

# --- (Дополнительно) Пример загрузки ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
try:
    loaded_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                               gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                               mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE)
    try:
        loaded_model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
        loaded_model.to(device); loaded_model.eval()
        print(f"Модель CNN-GRU успешно загружена из {MODEL_SAVE_PATH}")
    except FileNotFoundError: print(f"Файл модели {MODEL_SAVE_PATH} не найден.")
    except Exception as e: print(f"Ошибка при загрузке модели: {e}")
except NameError: print("Переменная 'input_dim' не определена.")
try:
    with open(SCALER_SAVE_PATH, 'rb') as f: loaded_scaler = pickle.load(f)
    print(f"Scaler успешно загружен из {SCALER_SAVE_PATH}")
    # Пример использования:
    # last_sequence = ... # ваши последние SEQUENCE_LENGTH дней признаков, shape (sequence_length, input_dim)
    # last_sequence_scaled = loaded_scaler.transform(last_sequence)
    # new_data_tensor = torch.tensor(last_sequence_scaled, dtype=torch.float32).unsqueeze(0).to(device) # Добавляем batch dimension
    # if 'loaded_model' in locals():
    #     with torch.no_grad(): prediction = loaded_model(new_data_tensor)
    #     print(f"Предсказание для следующего шага: {prediction.item()}")
except FileNotFoundError: print(f"Файл scaler'а {SCALER_SAVE_PATH} не найден.")
except Exception as e: print(f"Ошибка при загрузке scaler'а: {e}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
# >>> ИСПОЛЬЗУЕМ ADAMW <<<
# from torch.optim import Adam
from torch.optim import AdamW
from torch.utils.data import TensorDataset, DataLoader, Subset
# >>> ИСПОЛЬЗУЕМ OneCycleLR <<<
# from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.optim.lr_scheduler import OneCycleLR
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
model_save_name = 'cnngru_adamw_onecycle.pth' # Имя новой модели
scaler_save_name = 'cnngru_adamw_onecycle_scaler.pth' # Имя нового scaler'а
checkpoint_dir = os.path.join(base_path, 'checkpoints') # Директория для чекпоинтов
os.makedirs(checkpoint_dir, exist_ok=True) # Создаем директорию, если ее нет

target_col = 'Z_IMO_return' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ

SEQUENCE_LENGTH = 15 # Длина последовательности

# Параметры разделения и обучения
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 5
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 32

# >>> ПАРАМЕТРЫ ОБУЧЕНИЯ и ПЛАНИРОВЩИКА OneCycleLR <<<
# OneCycleLR сам управляет LR. Здесь задаем максимальный LR.
max_learning_rate = 0.008 # Можно попробовать немного агрессивнее
epochs = 200 # Кол-во эпох должно быть достаточным для полного цикла OneCycleLR

# >>> ПАРАМЕТРЫ РЕГУЛЯРИЗАЦИИ И ОПТИМИЗАТОРА <<<
# Вероятность Dropout (можно увеличить)
DROPOUT_RATE = 0.4
# Weight Decay (для AdamW)
WEIGHT_DECAY = 1e-4
# >>> Gradient Clipping <<<
MAX_GRAD_NORM = 1.0 # Максимальная норма градиента

# Параметры CNN-GRU (оставляем как было)
CNN_OUT_CHANNELS = 32
CNN_KERNEL_SIZE = 3
GRU_HIDDEN_SIZE = 64
GRU_NUM_LAYERS = 1 # Начнем с 1 слоя GRU после CNN
MLP_HIDDEN_SIZE = 32

# Параметры Early Stopping
EARLY_STOPPING_PATIENCE = 50 # Терпение
# SCHEDULER_PATIENCE / SCHEDULER_FACTOR НЕ НУЖНЫ для OneCycleLR

random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}"); print(f"Чекпоинты: {checkpoint_dir}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков (2D) ---
# >>> МАСШТАБИРУЕМ ЗДЕСЬ ОДИН РАЗ НА 2D ДАННЫХ <<<
scaler = StandardScaler()
X_scaled_2d = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled_2d = df_cleaned[target_col].values.reshape(-1, 1)
# Сохраняем этот scaler! Он нужен для новых данных.
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей (из УЖЕ масштабированных 2D данных) ---
def create_sequences(input_data_2d, target_data_2d, sequence_length):
    sequences, targets = [], []
    # Убедимся, что входные данные 2D и имеют одинаковую длину
    assert input_data_2d.shape[0] == target_data_2d.shape[0]
    assert input_data_2d.ndim == 2
    assert target_data_2d.ndim == 2

    for i in range(sequence_length, len(input_data_2d)):
        # Берем кусок из 2D массива по временной оси
        sequences.append(input_data_2d[i-sequence_length:i, :]) # Берем все признаки (:)
        targets.append(target_data_2d[i, :]) # Цель - один момент времени
    return np.array(sequences), np.array(targets) # sequences будет 3D, targets будет 2D

# >>> Создаем последовательности из масштабированных данных <<<
X_seq, y_seq = create_sequences(X_scaled_2d, y_unscaled_2d, SEQUENCE_LENGTH)
# Индексы соответствуют КОНЦУ каждой последовательности
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]

print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ (уже последовательностей) ---
print("-" * 30); print(f"Разделение последовательностей на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ...")
# Разделяем 3D последовательности X_seq, 2D цели y_seq и соответствующие индексы
X_temp, X_final_test, y_temp, y_final_test, temp_indices, final_test_indices = train_test_split(
    X_seq, y_seq, original_indices_seq, test_size=FINAL_TEST_SET_SIZE, random_state=random_seed, shuffle=True
)
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)


# --- 5. Определение МОДЕЛИ (CNN-GRU) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate
        # 1D CNN Слой: in_channels = input_dim, out_channels = cnn_out_channels
        # Вход: (N, C_in, L_in) = (batch_size, input_dim, sequence_length)
        self.conv1d = nn.Conv1d(in_channels=input_dim,
                                out_channels=cnn_out_channels,
                                kernel_size=kernel_size,
                                padding='same', # 'same' сохраняет длину последовательности
                                padding_mode='replicate') # Можно использовать replicate padding
        self.bn1 = nn.BatchNorm1d(cnn_out_channels) # BatchNorm применяется к каналам
        self.relu1 = nn.ReLU()
        self.dropout_cnn = nn.Dropout(dropout_rate)

        # GRU Слой: input_size = cnn_out_channels
        # Вход: (N, L, H_in) = (batch_size, sequence_length, cnn_out_channels)
        self.gru = nn.GRU(input_size=cnn_out_channels,
                          hidden_size=gru_hidden_dim,
                          num_layers=gru_num_layers,
                          batch_first=True,
                          dropout=dropout_rate if gru_num_layers > 1 else 0)

        # MLP Голова
        self.mlp = nn.Sequential(
            nn.Linear(gru_hidden_dim, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(mlp_hidden, output_dim)
        )

    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_dim)

        # --- CNN Часть ---
        # Меняем местами измерения для Conv1d: (N, C_in, L_in)
        x_cnn = x.permute(0, 2, 1)
        x_cnn = self.conv1d(x_cnn)
        x_cnn = self.bn1(x_cnn)
        x_cnn = self.relu1(x_cnn)
        x_cnn = self.dropout_cnn(x_cnn)
        # Меняем местами обратно для GRU: (N, L_out, C_out)
        x_gru_input = x_cnn.permute(0, 2, 1)

        # --- GRU Часть ---
        # Пропускаем через GRU
        out_gru, hn = self.gru(x_gru_input)

        # --- MLP Часть ---
        # Используем выход GRU с последнего временного шага: out_gru[:, -1, :]
        # Или последнее скрытое состояние: hn[-1]
        last_time_step_out = out_gru[:, -1, :]
        # last_time_step_out = hn[-1] # Альтернатива

        output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2] # Количество признаков в конце последовательности

# --- 6. Кросс-валидация ---
kf = KFold(n_splits=K_FOLDS, shuffle=True, random_state=random_seed)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (CNN-GRU + AdamW + OneCycleLR) ---")
for fold, (train_idx, val_idx) in enumerate(kf.split(X_temp, y_temp)):
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    # Подготовка данных фолда (уже последовательности, уже масштабированы)
    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]
    # Преобразуем в тензоры и переносим на устройство
    X_train_fold_tensor = torch.tensor(X_train_fold, dtype=torch.float32).to(device)
    y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold, dtype=torch.float32).to(device)
    y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)

    # DataLoader с drop_last=True для тренировки, False для валидации
    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor)
    # >>> drop_last=True для train_loader <<<
    train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor)
    val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False, drop_last=False)

    # Инициализация модели
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss()
    # >>> Оптимизатор AdamW <<<
    optimizer = AdamW(model.parameters(), lr=max_learning_rate, weight_decay=WEIGHT_DECAY) # Начальный LR здесь - это max_lr для OneCycleLR
    # >>> Планировщик OneCycleLR <<<
    steps_per_epoch = len(train_fold_loader)
    # Убедимся, что steps_per_epoch > 0
    if steps_per_epoch == 0:
        print(f"  Предупреждение: steps_per_epoch равен 0 в фолде {fold+1}. Пропуск фолда.")
        continue # Пропускаем этот фолд, если нет данных для обучения
    scheduler = OneCycleLR(optimizer, max_lr=max_learning_rate, epochs=epochs, steps_per_epoch=steps_per_epoch)

    # Обучение фолда с Early Stopping
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1} (модель CNN-GRU)...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y)
            loss.backward()
            # >>> Gradient Clipping <<<
            torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)
            optimizer.step()
            # >>> OneCycleLR шаг ПОСЛЕ каждого батча <<<
            scheduler.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        # Обработка случая, если не было ни одного батча
        if batches_processed_train == 0:
            print(f"    Предупреждение: Ни одного батча не обработано в эпохе {epoch+1}, фолд {fold+1}.")
            continue # Переходим к следующей эпохе
        avg_train_loss = train_loss_epoch / batches_processed_train

        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        with torch.no_grad():
            for batch_X_val, batch_y_val in val_fold_loader:
                outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
        avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')

        if (epoch + 1) % 1 == 0: print(f'    Фолд {fold+1}, Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        # Early Stopping
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено на эпохе {epoch+1} (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель фолда.")
    model.eval()
    with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
    mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
    print(f"  Метрики валидации фолда {fold+1} (лучшая): MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
    fold_mses.append(mse_fold); fold_r2s.append(r2_fold)

# --- Результаты кросс-валидации ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU + AdamW + OneCycleLR) ---")
# Проверяем, есть ли результаты, прежде чем считать среднее
if fold_mses:
    print(f"Среднее MSE по фолдам: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})")
    print(f"Среднее R² по фолдам: {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
    print(f"Среднее количество 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")
else:
    print("Недостаточно данных для проведения кросс-валидации.")


# --- 7. Обучение ФИНАЛЬНОЙ модели ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU + AdamW + OneCycleLR) ---")
# Подготовка данных
X_train_final_seq, X_val_final_seq, y_train_final_seq, y_val_final_seq = train_test_split(X_temp, y_temp, test_size=VAL_SPLIT_FOR_FINAL, random_state=random_seed, shuffle=True)
X_train_final_tensor = torch.tensor(X_train_final_seq, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
X_val_final_tensor = torch.tensor(X_val_final_seq, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
# DataLoader с drop_last=True для тренировки, False для валидации
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor); final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False, drop_last=False)

# Инициализация модели
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss()
# Оптимизатор AdamW
final_optimizer = AdamW(final_model.parameters(), lr=max_learning_rate, weight_decay=WEIGHT_DECAY)
# Планировщик OneCycleLR
final_steps_per_epoch = len(final_train_loader)
if final_steps_per_epoch == 0: raise ValueError("Нет данных для финального обучения (final_steps_per_epoch=0).")
final_scheduler = OneCycleLR(final_optimizer, max_lr=max_learning_rate, epochs=epochs, steps_per_epoch=final_steps_per_epoch)

# Обучение финальной модели с Early Stopping и чекпоинтами
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения (модель CNN-GRU)...")
start_time_final = time.time()
for epoch in range(epochs):
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y)
        loss.backward()
        # Gradient Clipping
        torch.nn.utils.clip_grad_norm_(final_model.parameters(), MAX_GRAD_NORM)
        final_optimizer.step()
        # OneCycleLR шаг ПОСЛЕ каждого батча
        final_scheduler.step()
        train_loss_epoch += loss.item(); batches_processed_train += 1
    avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in final_val_loader:
            outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
            val_loss_epoch += loss_val.item(); batches_processed_val += 1
    avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')

    if (epoch + 1) % 1 == 0: print(f'    Финал. обуч., Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')

    # Early Stopping + Сохранение лучшего чекпоинта
    if avg_val_loss < best_val_loss_final:
        best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
        best_checkpoint_path = os.path.join(checkpoint_dir, f'best_model_final.pth') # Упрощенное имя
        try:
            torch.save({'epoch': best_epoch_final, 'model_state_dict': best_model_state_final,
                        'optimizer_state_dict': final_optimizer.state_dict(), 'scheduler_state_dict': final_scheduler.state_dict(),
                        'loss': best_val_loss_final,}, best_checkpoint_path)
            # print(f"    Сохранен лучший чекпоинт: {best_checkpoint_path}") # Можно раскомментировать для дебага
        except Exception as e: print(f"    Ошибка сохранения лучшего чекпоинта: {e}")
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE:
            print(f"    Ранняя остановка фин. обучения! Нет улучшения Val Loss за {EARLY_STOPPING_PATIENCE} эпох.")
            break
    # Сохраняем периодический чекпоинт
    if (epoch + 1) % 50 == 0:
         checkpoint_path_epoch = os.path.join(checkpoint_dir, f'epoch_{epoch+1}_final.pth') # Упрощенное имя
         try:
             torch.save({'epoch': epoch + 1, 'model_state_dict': final_model.state_dict(),
                         'optimizer_state_dict': final_optimizer.state_dict(), 'scheduler_state_dict': final_scheduler.state_dict(),
                         'loss': avg_val_loss,}, checkpoint_path_epoch)
             # print(f"    Сохранен чекпоинт эпохи {epoch+1}: {checkpoint_path_epoch}") # Можно раскомментировать
         except Exception as e: print(f"    Ошибка сохранения чекпоинта эпохи {epoch+1}: {e}")


end_time_final = time.time(); print(f"Финальное обучение завершено на эпохе {epoch+1} (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели (из чекпоинта) ---
# Загружаем лучшую модель из сохраненного чекпоинта для финальной оценки
final_model_for_eval = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                           gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                           mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
if best_model_state_final:
    try:
        best_checkpoint_path = os.path.join(checkpoint_dir, f'best_model_final.pth')
        checkpoint = torch.load(best_checkpoint_path, map_location=device)
        final_model_for_eval.load_state_dict(checkpoint['model_state_dict'])
        final_model_for_eval.eval() # Переводим в режим оценки
        # Сохраняем только state_dict лучшей модели в основной файл
        torch.save(checkpoint['model_state_dict'], MODEL_SAVE_PATH)
        print(f"Лучшая финальная модель (эпоха {checkpoint['epoch']}) загружена из чекпоинта и сохранена в: {MODEL_SAVE_PATH}")
    except Exception as e:
        print(f"Ошибка загрузки/сохранения лучшей финальной модели из чекпоинта: {e}. Используется последняя обученная модель.")
        final_model_for_eval = final_model # Используем модель, которая осталась после обучения
        final_model_for_eval.eval() # Переводим ее в eval
else:
    print("Предупреждение: Не найдено лучшего состояния финальной модели. Используется последняя обученная модель.")
    final_model_for_eval = final_model # Используем модель, которая осталась после обучения
    final_model_for_eval.eval() # Переводим ее в eval

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ + Сравнение с SMA(5) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
# ВАЖНО: X_final_test УЖЕ содержит ПОСЛЕДОВАТЕЛЬНОСТИ масштабированных данных
X_final_test_tensor_eval = torch.tensor(X_final_test, dtype=torch.float32).to(device)
with torch.no_grad(): y_pred_nn = final_model_for_eval(X_final_test_tensor_eval).cpu().numpy()

# Получение предсказаний базовой модели SMA(5)
y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
# Обработка NaN в начале SMA(5)
mean_y_test = np.mean(y_final_test)
nan_mask_sma5 = np.isnan(y_pred_sma5)
if np.any(nan_mask_sma5):
    print(f"Обнаружено {np.sum(nan_mask_sma5)} NaN в прогнозах SMA(5), заполняются средним значением теста ({mean_y_test:.4f}).")
    y_pred_sma5[nan_mask_sma5] = mean_y_test
else:
    print("NaN в прогнозах SMA(5) не обнаружено.")

# Расчет метрик
mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
ssr_model = mse_final_nn * len(y_final_test)
mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0

print("\n--- Результаты оценки (CNN-GRU + AdamW + OneCycleLR) ---")
print(f"MSE NN:               {mse_final_nn:.4f}")
print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}")
print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")
print("---")
print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}")
print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}") # <<< СРАВНЕНИЕ С SMA(5)

# --- 10. Визуализация ---
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
plt.plot([y_final_test.min(), y_final_test.max()], [y_final_test.min(), y_final_test.max()], 'r--', lw=2, label='Идеал')
plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест, CNN-GRU)\nЦель: {target_col}")
plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten(), 'Predicted_SMA5': y_pred_sma5.flatten()}, index=final_test_indices).sort_index()
n_points_to_plot = min(100, len(results_final_test_df))
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)\nЦель: {target_col}")
plt.xticks(rotation=45), plt.legend(), plt.grid(True); plt.tight_layout(); plt.show()
plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (CNN-GRU + OneCycleLR)"); plt.legend(), plt.grid(True); plt.show()

# --- (Дополнительно) Пример загрузки ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
try:
    loaded_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                               gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                               mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE)
    try:
        checkpoint_state_dict = torch.load(MODEL_SAVE_PATH, map_location=device)
        loaded_model.load_state_dict(checkpoint_state_dict)
        loaded_model.to(device); loaded_model.eval()
        print(f"Модель CNN-GRU успешно загружена из {MODEL_SAVE_PATH}")
    except FileNotFoundError: print(f"Файл модели {MODEL_SAVE_PATH} не найден.")
    except Exception as e: print(f"Ошибка при загрузке модели: {e}")
except NameError: print("Переменная 'input_dim' не определена.")
try:
    with open(SCALER_SAVE_PATH, 'rb') as f: loaded_scaler = pickle.load(f); print(f"Scaler успешно загружен из {SCALER_SAVE_PATH}")
    # Пример использования:
    # last_N_days_raw_2d_data = ... # shape (SEQUENCE_LENGTH, input_dim)
    # if last_N_days_raw_2d_data.shape[0] == SEQUENCE_LENGTH and last_N_days_raw_2d_data.shape[1] == input_dim:
    #    scaled_sequence_2d = loaded_scaler.transform(last_N_days_raw_2d_data)
    #    scaled_sequence_3d = scaled_sequence_2d[np.newaxis, :, :]
    #    new_data_tensor = torch.tensor(scaled_sequence_3d, dtype=torch.float32).to(device)
    #    if 'loaded_model' in locals() and not loaded_model.training:
    #        with torch.no_grad(): prediction = loaded_model(new_data_tensor)
    #        print(f"Предсказание для следующего шага: {prediction.item()}")
    #    else: print("Модель не загружена / не в eval mode.")
    # else: print("Некорректная форма входных данных.")
except FileNotFoundError: print(f"Файл scaler'а {SCALER_SAVE_PATH} не найден.")
except Exception as e: print(f"Ошибка при загрузке scaler'а: {e}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
# >>> НУЖЕН ДЛЯ WeightNorm <<<
from torch.nn.utils import weight_norm
# >>> ОПТИМИЗАТОР И ПЛАНИРОВЩИК ОСТАВЛЯЕМ ТЕ ЖЕ <<<
from torch.optim import AdamW
from torch.optim.lr_scheduler import OneCycleLR
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
# >>> НОВЫЕ ИМЕНА для TCN модели <<<
model_save_name = 'tcn_model_improved.pth'
scaler_save_name = 'tcn_scaler_improved.pth'
checkpoint_dir = os.path.join(base_path, 'checkpoints_tcn') # Отдельная папка
os.makedirs(checkpoint_dir, exist_ok=True)

target_col = 'Z_IMO_return' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ

SEQUENCE_LENGTH = 20 # TCN хорошо работают с длинными последовательностями

# Параметры разделения и обучения
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 2
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 64
max_learning_rate = 0.005 # LR для OneCycleLR
epochs = 200 # Макс. кол-во эпох

# >>> ПАРАМЕТРЫ TCN МОДЕЛИ <<<
TCN_NUM_CHANNELS = [32, 32, 32, 64] # Количество каналов (фильтров) на каждом уровне TCN
TCN_KERNEL_SIZE = 5       # Размер ядра свертки
TCN_DROPOUT = 0.3             # Dropout внутри TCN и после

# Параметры регуляризации и оптимизации
WEIGHT_DECAY = 1e-4
MAX_GRAD_NORM = 1.0
EARLY_STOPPING_PATIENCE = 60 # Немного увеличим терпение для TCN

random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}"); print(f"Чекпоинты: {checkpoint_dir}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков (2D) ---
scaler = StandardScaler()
X_scaled_2d = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled_2d = df_cleaned[target_col].values.reshape(-1, 1)
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей ---
def create_sequences(input_data_2d, target_data_2d, sequence_length):
    sequences, targets = [], []
    assert input_data_2d.shape[0] == target_data_2d.shape[0]; assert input_data_2d.ndim == 2; assert target_data_2d.ndim == 2
    for i in range(sequence_length, len(input_data_2d)):
        sequences.append(input_data_2d[i-sequence_length:i, :])
        targets.append(target_data_2d[i, :])
    return np.array(sequences), np.array(targets)
X_seq, y_seq = create_sequences(X_scaled_2d, y_unscaled_2d, SEQUENCE_LENGTH)
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ ---
print("-" * 30); print(f"Разделение последовательностей на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ...")
X_temp, X_final_test, y_temp, y_final_test, temp_indices, final_test_indices = train_test_split(
    X_seq, y_seq, original_indices_seq, test_size=FINAL_TEST_SET_SIZE, random_state=random_seed, shuffle=True
)
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)


# --- 5. Определение МОДЕЛИ (TCN) ---
# Вспомогательный модуль для обрезки паддинга в причинных свертках
class Chomp1d(nn.Module):
    def __init__(self, chomp_size):
        super(Chomp1d, self).__init__()
        self.chomp_size = chomp_size
    def forward(self, x):
        return x[:, :, :-self.chomp_size].contiguous()

# Основной блок TCN с остаточными соединениями
class TemporalBlock(nn.Module):
    def __init__(self, n_inputs, n_outputs, kernel_size, stride, dilation, padding, dropout=0.2):
        super(TemporalBlock, self).__init__()
        # Первая свертка + нормализация + активация + dropout
        self.conv1 = weight_norm(nn.Conv1d(n_inputs, n_outputs, kernel_size,
                                           stride=stride, padding=padding, dilation=dilation))
        self.chomp1 = Chomp1d(padding) # Обрезаем паддинг для причинности
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(dropout)

        # Вторая свертка + нормализация + активация + dropout
        self.conv2 = weight_norm(nn.Conv1d(n_outputs, n_outputs, kernel_size,
                                           stride=stride, padding=padding, dilation=dilation))
        self.chomp2 = Chomp1d(padding)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(dropout)

        # Сборка слоев в Sequential
        self.net = nn.Sequential(self.conv1, self.chomp1, self.relu1, self.dropout1,
                                 self.conv2, self.chomp2, self.relu2, self.dropout2)

        # 1x1 свертка для согласования размерности в остаточном соединении, если нужно
        self.downsample = nn.Conv1d(n_inputs, n_outputs, 1) if n_inputs != n_outputs else None
        self.relu = nn.ReLU()
        self.init_weights()

    def init_weights(self):
        # Инициализация весов (может помочь сходимости)
        self.conv1.weight.data.normal_(0, 0.01)
        self.conv2.weight.data.normal_(0, 0.01)
        if self.downsample is not None:
            self.downsample.weight.data.normal_(0, 0.01)

    def forward(self, x):
        out = self.net(x)
        # Остаточное соединение
        res = x if self.downsample is None else self.downsample(x)
        return self.relu(out + res)

# Сама TCN модель
class TCNModel(nn.Module):
    def __init__(self, input_dim, output_dim, num_channels, kernel_size, dropout):
        super(TCNModel, self).__init__()
        layers = []
        num_levels = len(num_channels)
        for i in range(num_levels):
            dilation_size = 2 ** i # Экспоненциальное увеличение dilation
            in_channels = input_dim if i == 0 else num_channels[i-1]
            out_channels = num_channels[i]
            # Рассчитываем паддинг для причинной свертки
            padding = (kernel_size - 1) * dilation_size
            layers += [TemporalBlock(in_channels, out_channels, kernel_size, stride=1, dilation=dilation_size,
                                     padding=padding, dropout=dropout)]

        self.network = nn.Sequential(*layers)
        # Финальный полносвязный слой берет выход последнего шага TCN
        self.linear = nn.Linear(num_channels[-1], output_dim)

    def forward(self, x):
        # Вход TCN: (N, C_in, L_in) = (batch_size, input_dim, sequence_length)
        x = x.permute(0, 2, 1) # Меняем местами признаки и длину
        y = self.network(x)
        # Берем выход с последнего временного шага: y[:, :, -1]
        # shape: (batch_size, num_channels[-1])
        o = self.linear(y[:, :, -1])
        return o

input_dim = X_seq.shape[2] # Количество признаков

# --- 6. Кросс-валидация ---
kf = KFold(n_splits=K_FOLDS, shuffle=True, random_state=random_seed)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (TCN + AdamW + OneCycleLR) ---")
for fold, (train_idx, val_idx) in enumerate(kf.split(X_temp, y_temp)):
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    # Подготовка данных
    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]; y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]
    X_train_fold_tensor = torch.tensor(X_train_fold, dtype=torch.float32).to(device); y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold, dtype=torch.float32).to(device); y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)
    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False, drop_last=False)

    # >>> Инициализация TCN модели <<<
    model = TCNModel(input_dim=input_dim, output_dim=1, num_channels=TCN_NUM_CHANNELS,
                     kernel_size=TCN_KERNEL_SIZE, dropout=TCN_DROPOUT).to(device)
    criterion = nn.MSELoss()
    optimizer = AdamW(model.parameters(), lr=max_learning_rate, weight_decay=WEIGHT_DECAY)
    steps_per_epoch = len(train_fold_loader)
    if steps_per_epoch == 0: print(f"  Предупреждение: steps_per_epoch=0 в фолде {fold+1}. Пропуск."); continue
    scheduler = OneCycleLR(optimizer, max_lr=max_learning_rate, epochs=epochs, steps_per_epoch=steps_per_epoch)

    # Обучение фолда
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1} (модель TCN)...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y)
            loss.backward(); torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM); optimizer.step(); scheduler.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        if batches_processed_train == 0: print(f"    Предупреждение: Нет батчей в эпохе {epoch+1}."); continue
        avg_train_loss = train_loss_epoch / batches_processed_train

        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        with torch.no_grad():
            for batch_X_val, batch_y_val in val_fold_loader:
                outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
        avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')

        if (epoch + 1) % 1 == 0: print(f'    Фолд {fold+1}, Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено на эпохе {epoch+1} (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель фолда.")
    model.eval()
    with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
    mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
    print(f"  Метрики валидации фолда {fold+1} (лучшая): MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
    fold_mses.append(mse_fold); fold_r2s.append(r2_fold)

# --- Результаты кросс-валидации ---
print("\n--- Сводные результаты К-кратной кросс-валидации (TCN + AdamW + OneCycleLR) ---")
if fold_mses:
    print(f"Среднее MSE по фолдам: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})")
    print(f"Среднее R² по фолдам: {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
    print(f"Среднее количество 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")
else: print("Кросс-валидация не дала результатов.")


# --- 7. Обучение ФИНАЛЬНОЙ модели ---
print("\n--- Обучение финальной модели на ТЕМП данных (TCN + AdamW + OneCycleLR) ---")
# Подготовка данных
X_train_final_seq, X_val_final_seq, y_train_final_seq, y_val_final_seq = train_test_split(X_temp, y_temp, test_size=VAL_SPLIT_FOR_FINAL, random_state=random_seed, shuffle=True)
X_train_final_tensor = torch.tensor(X_train_final_seq, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
X_val_final_tensor = torch.tensor(X_val_final_seq, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor); final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False, drop_last=False)

# >>> Инициализация TCN финальной модели <<<
final_model = TCNModel(input_dim=input_dim, output_dim=1, num_channels=TCN_NUM_CHANNELS,
                     kernel_size=TCN_KERNEL_SIZE, dropout=TCN_DROPOUT).to(device)
final_criterion = nn.MSELoss()
final_optimizer = AdamW(final_model.parameters(), lr=max_learning_rate, weight_decay=WEIGHT_DECAY)
final_steps_per_epoch = len(final_train_loader)
if final_steps_per_epoch == 0: raise ValueError("Нет данных для финального обучения.")
final_scheduler = OneCycleLR(final_optimizer, max_lr=max_learning_rate, epochs=epochs, steps_per_epoch=final_steps_per_epoch)

# Обучение финальной модели
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения (модель TCN)...")
start_time_final = time.time()
for epoch in range(epochs):
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y)
        loss.backward(); torch.nn.utils.clip_grad_norm_(final_model.parameters(), MAX_GRAD_NORM); final_optimizer.step(); final_scheduler.step()
        train_loss_epoch += loss.item(); batches_processed_train += 1
    if batches_processed_train == 0: continue
    avg_train_loss = train_loss_epoch / batches_processed_train; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in final_val_loader:
            outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
            val_loss_epoch += loss_val.item(); batches_processed_val += 1
    avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')

    if (epoch + 1) % 1 == 0: print(f'    Финал. обуч., Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')
    if avg_val_loss < best_val_loss_final:
        best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
        best_checkpoint_path = os.path.join(checkpoint_dir, f'best_model_tcn_final.pth')
        try:
            torch.save({'epoch': best_epoch_final, 'model_state_dict': best_model_state_final,
                        'optimizer_state_dict': final_optimizer.state_dict(), 'scheduler_state_dict': final_scheduler.state_dict(),
                        'loss': best_val_loss_final,}, best_checkpoint_path)
            # print(f"    Сохранен лучший TCN чекпоинт.")
        except Exception as e: print(f"    Ошибка сохранения TCN чекпоинта: {e}")
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка фин. обучения!"); break
    if (epoch + 1) % 50 == 0:
         checkpoint_path_epoch = os.path.join(checkpoint_dir, f'epoch_{epoch+1}_tcn_final.pth')
         try:
             torch.save({'epoch': epoch + 1, 'model_state_dict': final_model.state_dict(),
                         'optimizer_state_dict': final_optimizer.state_dict(), 'scheduler_state_dict': final_scheduler.state_dict(),
                         'loss': avg_val_loss,}, checkpoint_path_epoch)
             # print(f"    Сохранен чекпоинт TCN эпохи {epoch+1}.")
         except Exception as e: print(f"    Ошибка сохранения TCN чекпоинта эпохи {epoch+1}: {e}")

end_time_final = time.time(); print(f"Финальное обучение завершено на эпохе {epoch+1} (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели ---
final_model_for_eval = TCNModel(input_dim=input_dim, output_dim=1, num_channels=TCN_NUM_CHANNELS,
                     kernel_size=TCN_KERNEL_SIZE, dropout=TCN_DROPOUT).to(device)
if best_model_state_final:
    try:
        best_checkpoint_path = os.path.join(checkpoint_dir, f'best_model_tcn_final.pth')
        checkpoint = torch.load(best_checkpoint_path, map_location=device)
        final_model_for_eval.load_state_dict(checkpoint['model_state_dict'])
        final_model_for_eval.eval()
        torch.save(checkpoint['model_state_dict'], MODEL_SAVE_PATH)
        print(f"Лучшая TCN модель (эпоха {checkpoint['epoch']}) загружена и сохранена в: {MODEL_SAVE_PATH}")
    except Exception as e: print(f"Ошибка загрузки/сохранения лучшей TCN модели: {e}. Используется последняя."); final_model_for_eval = final_model; final_model_for_eval.eval()
else: print("Предупреждение: Лучшей TCN модели нет. Используется последняя."); final_model_for_eval = final_model; final_model_for_eval.eval()

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ + Сравнение с SMA(5) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели (TCN) на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
final_model_for_eval.eval()
X_final_test_tensor_eval = torch.tensor(X_final_test, dtype=torch.float32).to(device)
with torch.no_grad(): y_pred_nn = final_model_for_eval(X_final_test_tensor_eval).cpu().numpy()
y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
else: print("NaN в SMA(5) не обнаружено.")

mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
ssr_model = mse_final_nn * len(y_final_test)
mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0

print("\n--- Результаты оценки (TCN + AdamW + OneCycleLR) ---")
print(f"MSE TCN:               {mse_final_nn:.4f}")
print(f"RMSE TCN:             {np.sqrt(mse_final_nn):.4f}")
print(f"MAE TCN:              {mean_absolute_error(y_final_test, y_pred_nn):.4f}")
print(f"R² TCN (vs Mean):     {r2_final_nn:.4f}")
print("---")
print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}")
print(f"R² TCN (vs SMA(5)):   {r2_vs_sma5:.4f}") # <<< СРАВНЕНИЕ С SMA(5)

# --- 10. Визуализация ---
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание TCN vs Реальность')
plt.plot([y_final_test.min(), y_final_test.max()], [y_final_test.min(), y_final_test.max()], 'r--', lw=2, label='Идеал')
plt.xlabel("Реальные"), plt.ylabel("Предсказанные (TCN)"); plt.title(f"Сравнение (Фин. тест, TCN)\nЦель: {target_col}")
plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten(), 'Predicted_SMA5': y_pred_sma5.flatten()}, index=final_test_indices).sort_index()
n_points_to_plot = min(100, len(results_final_test_df))
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания TCN', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)\nЦель: {target_col}")
plt.xticks(rotation=45), plt.legend(), plt.grid(True); plt.tight_layout(); plt.show()
plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (TCN + OneCycleLR)"); plt.legend(), plt.grid(True); plt.show()

# --- (Дополнительно) Пример загрузки ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
try:
    loaded_model = TCNModel(input_dim=input_dim, output_dim=1, num_channels=TCN_NUM_CHANNELS,
                             kernel_size=TCN_KERNEL_SIZE, dropout=TCN_DROPOUT)
    try:
        checkpoint_state_dict = torch.load(MODEL_SAVE_PATH, map_location=device)
        loaded_model.load_state_dict(checkpoint_state_dict)
        loaded_model.to(device); loaded_model.eval()
        print(f"Модель TCN успешно загружена из {MODEL_SAVE_PATH}")
    except FileNotFoundError: print(f"Файл модели {MODEL_SAVE_PATH} не найден.")
    except Exception as e: print(f"Ошибка при загрузке модели: {e}")
except NameError: print("Переменная 'input_dim' не определена.")
try:
    with open(SCALER_SAVE_PATH, 'rb') as f: loaded_scaler = pickle.load(f); print(f"Scaler успешно загружен из {SCALER_SAVE_PATH}")
    # Пример использования:
    # last_N_days_raw_2d_data = ... # shape (SEQUENCE_LENGTH, input_dim)
    # if last_N_days_raw_2d_data.shape[0] == SEQUENCE_LENGTH and last_N_days_raw_2d_data.shape[1] == input_dim:
    #    scaled_sequence_2d = loaded_scaler.transform(last_N_days_raw_2d_data)
    #    scaled_sequence_3d = scaled_sequence_2d[np.newaxis, :, :]
    #    new_data_tensor = torch.tensor(scaled_sequence_3d, dtype=torch.float32).to(device)
    #    if 'loaded_model' in locals() and not loaded_model.training:
    #        with torch.no_grad(): prediction = loaded_model(new_data_tensor)
    #        print(f"Предсказание для следующего шага: {prediction.item()}")
    #    else: print("Модель не загружена / не в eval mode.")
    # else: print("Некорректная форма входных данных.")
except FileNotFoundError: print(f"Файл scaler'а {SCALER_SAVE_PATH} не найден.")
except Exception as e: print(f"Ошибка при загрузке scaler'а: {e}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, Subset
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
# >>> НОВЫЕ ИМЕНА для CNN-GRU модели <<<
model_save_name = 'cnn_gru_nn_improved.pth'
scaler_save_name = 'scaler_cnn_gru_improved.pth'

target_col = 'IMO_fwd_log_ret_3d' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ

# Параметры последовательности
SEQUENCE_LENGTH = 20 # Можно попробовать немного увеличить окно

# Параметры разделения и обучения
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 3
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 64
learning_rate = 0.001
epochs = 220 # Макс. кол-во эпох

# >>> ПАРАМЕТРЫ CNN-GRU МОДЕЛИ <<<
CNN_OUT_CHANNELS = 32 # Количество фильтров (признаков) после CNN
CNN_KERNEL_SIZE = 3   # Размер окна свертки (например, 3 или 5)
GRU_HIDDEN_SIZE = 64 # Размер скрытого состояния GRU
GRU_NUM_LAYERS = 1   # Попробуем 1 слой GRU после CNN
MLP_HIDDEN_SIZE = 32 # Размер скрытого слоя в MLP после GRU
DROPOUT_RATE = 0.3   # Dropout

# Параметры улучшений
EARLY_STOPPING_PATIENCE = 500
SCHEDULER_PATIENCE = 100
SCHEDULER_FACTOR = 0.1
WEIGHT_DECAY = 1e-5

random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled = df_cleaned[target_col].values.reshape(-1, 1)
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей ---
def create_sequences(input_data, target_data, sequence_length):
    sequences, targets = [], []
    for i in range(sequence_length, len(input_data)):
        sequences.append(input_data[i-sequence_length:i])
        targets.append(target_data[i])
    return np.array(sequences), np.array(targets)
X_seq, y_seq = create_sequences(X_scaled, y_unscaled, SEQUENCE_LENGTH)
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ ---
print("-" * 30); print(f"Разделение последовательностей на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ...")
X_temp, X_final_test, y_temp, y_final_test, temp_indices, final_test_indices = train_test_split(
    X_seq, y_seq, original_indices_seq, test_size=FINAL_TEST_SET_SIZE, random_state=random_seed, shuffle=True
)
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)


# --- 5. Определение НОВОЙ МОДЕЛИ (CNN-GRU) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate

        # 1D CNN Слой
        # Вход: (N, C_in, L_in) = (batch_size, input_dim, sequence_length)
        # Выход: (N, C_out, L_out) = (batch_size, cnn_out_channels, sequence_length) (с padding='same')
        # Нам нужно сначала поменять местами sequence_length и input_dim
        self.conv1d = nn.Conv1d(in_channels=input_dim,
                                out_channels=cnn_out_channels,
                                kernel_size=kernel_size,
                                padding='same') # 'same' сохраняет длину последовательности
        self.bn1 = nn.BatchNorm1d(cnn_out_channels) # BatchNorm применяется к каналам
        self.relu1 = nn.ReLU()
        self.dropout_cnn = nn.Dropout(dropout_rate)

        # GRU Слой
        # Вход: (N, L, H_in) = (batch_size, sequence_length, cnn_out_channels)
        # Выход hn: (num_layers, N, H_out) = (gru_num_layers, batch_size, gru_hidden_dim)
        # Нам нужно снова поменять местами измерения после CNN
        self.gru = nn.GRU(input_size=cnn_out_channels, # Входной размер GRU = выход CNN
                          hidden_size=gru_hidden_dim,
                          num_layers=gru_num_layers,
                          batch_first=True, # Ожидаем (batch, seq, feature)
                          dropout=dropout_rate if gru_num_layers > 1 else 0)

        # MLP Голова
        self.mlp = nn.Sequential(
            nn.Linear(gru_hidden_dim, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(mlp_hidden, output_dim)
        )

    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_dim)

        # --- CNN Часть ---
        # Меняем местами измерения для Conv1d: (N, C_in, L_in)
        x_cnn = x.permute(0, 2, 1)
        x_cnn = self.conv1d(x_cnn)
        x_cnn = self.bn1(x_cnn)
        x_cnn = self.relu1(x_cnn)
        x_cnn = self.dropout_cnn(x_cnn)
        # Меняем местами обратно для GRU: (N, L_out, C_out)
        x_gru_input = x_cnn.permute(0, 2, 1)

        # --- GRU Часть ---
        # Инициализация скрытого состояния (не обязательно, GRU сделает это по умолчанию)
        # h0 = torch.zeros(self.gru.num_layers, x.size(0), self.gru.hidden_size).to(x.device)
        # out_gru, hn = self.gru(x_gru_input, h0)
        out_gru, hn = self.gru(x_gru_input) # Используем выход последнего шага

        # --- MLP Часть ---
        # Берем выход GRU с последнего временного шага
        # out_gru shape: (batch_size, sequence_length, gru_hidden_dim)
        last_time_step_out = out_gru[:, -1, :]
        # Или можно взять последнее скрытое состояние: hn[-1] для последнего слоя
        # last_time_step_out = hn[-1] # shape: (batch_size, gru_hidden_dim)

        output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2] # Количество признаков

# --- 6. Кросс-валидация ---
kf = KFold(n_splits=K_FOLDS, shuffle=True, random_state=random_seed)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
for fold, (train_idx, val_idx) in enumerate(kf.split(X_temp, y_temp)):
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    # Подготовка данных
    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]
    X_train_fold_tensor = torch.tensor(X_train_fold, dtype=torch.float32).to(device)
    y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold, dtype=torch.float32).to(device)
    y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)
    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True)
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False)

    # >>> Инициализация CNN-GRU модели <<<
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

    # Обучение фолда
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1} (модель CNN-GRU)...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y)
            loss.backward(); optimizer.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0
        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        with torch.no_grad():
            for batch_X_val, batch_y_val in val_fold_loader:
                outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
        avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
        if (epoch + 1) % 1 == 0: print(f'    Фолд {fold+1}, Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        scheduler.step(avg_val_loss)
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено на эпохе {epoch+1} (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель фолда.")
    model.eval()
    with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
    mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
    print(f"  Метрики валидации фолда {fold+1} (лучшая): MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
    fold_mses.append(mse_fold); fold_r2s.append(r2_fold)

# --- Результаты кросс-валидации ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
print(f"Среднее MSE по фолдам: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})")
print(f"Среднее R² по фолдам: {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
print(f"Среднее количество 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")

# --- 7. Обучение ФИНАЛЬНОЙ модели ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU NN + Улучшения) ---")
# Подготовка данных
X_train_final_seq, X_val_final_seq, y_train_final_seq, y_val_final_seq = train_test_split(X_temp, y_temp, test_size=VAL_SPLIT_FOR_FINAL, random_state=random_seed, shuffle=True)
X_train_final_tensor = torch.tensor(X_train_final_seq, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
X_val_final_tensor = torch.tensor(X_val_final_seq, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor); final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True)
final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False)

# >>> Инициализация CNN-GRU финальной модели <<<
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss()
final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
final_scheduler = ReduceLROnPlateau(final_optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

# Обучение финальной модели
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения (модель CNN-GRU)...")
start_time_final = time.time()
for epoch in range(epochs):
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y)
        loss.backward(); final_optimizer.step()
        train_loss_epoch += loss.item(); batches_processed_train += 1
    avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in final_val_loader:
            outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
            val_loss_epoch += loss_val.item(); batches_processed_val += 1
    avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
    if (epoch + 1) % 1 == 0: print(f'    Финал. обуч., Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')
    final_scheduler.step(avg_val_loss)
    if avg_val_loss < best_val_loss_final: best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка фин. обучения!"); break
end_time_final = time.time(); print(f"Финальное обучение завершено на эпохе {epoch+1} (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели ---
if best_model_state_final:
    try:
        torch.save(best_model_state_final, MODEL_SAVE_PATH); print(f"Лучшая финальная модель ({best_epoch_final}) сохранена: {MODEL_SAVE_PATH}")
        final_model.load_state_dict(best_model_state_final)
    except Exception as e: print(f"Ошибка сохранения/загрузки: {e}")
else: print("Предупреждение: Используется последняя модель.")

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ + Сравнение с SMA(5) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
final_model.eval()
# ВАЖНО: X_final_test УЖЕ содержит ПОСЛЕДОВАТЕЛЬНОСТИ масштабированных данных
X_final_test_tensor_eval = torch.tensor(X_final_test, dtype=torch.float32).to(device)
with torch.no_grad(): y_pred_nn = final_model(X_final_test_tensor_eval).cpu().numpy()
# Получение предсказаний SMA(5) для соответствующих дат КОНЦА последовательностей
y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
else: print("NaN в SMA(5) не обнаружено.")

# Расчет метрик
mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
ssr_model = mse_final_nn * len(y_final_test)
mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0

print("\n--- Результаты оценки (CNN-GRU NN + Улучшения) ---")
print(f"MSE NN:               {mse_final_nn:.4f}")
print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}")
print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")
print("---")
print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}")
print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}") # <<< СРАВНЕНИЕ С SMA(5)

# --- 10. Визуализация ---
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
plt.plot([y_final_test.min(), y_final_test.max()], [y_final_test.min(), y_final_test.max()], 'r--', lw=2, label='Идеал')
plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест, CNN-GRU)\nЦель: {target_col}")
plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten(), 'Predicted_SMA5': y_pred_sma5.flatten()}, index=final_test_indices).sort_index()
n_points_to_plot = min(100, len(results_final_test_df))
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)\nЦель: {target_col}")
plt.xticks(rotation=45), plt.legend(), plt.grid(True); plt.tight_layout(); plt.show()
plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (CNN-GRU NN)"); plt.legend(), plt.grid(True); plt.show()

# --- (Дополнительно) Пример загрузки ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
try:
    loaded_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                               gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                               mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE)
    try:
        loaded_model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
        loaded_model.to(device); loaded_model.eval()
        print(f"Модель CNN-GRU успешно загружена из {MODEL_SAVE_PATH}")
    except FileNotFoundError: print(f"Файл модели {MODEL_SAVE_PATH} не найден.")
    except Exception as e: print(f"Ошибка при загрузке модели: {e}")
except NameError: print("Переменная 'input_dim' не определена.")
try:
    with open(SCALER_SAVE_PATH, 'rb') as f: loaded_scaler = pickle.load(f)
    print(f"Scaler успешно загружен из {SCALER_SAVE_PATH}")
    # Пример использования:
    # last_sequence = ... # ваши последние SEQUENCE_LENGTH дней признаков, shape (sequence_length, input_dim)
    # last_sequence_scaled = loaded_scaler.transform(last_sequence)
    # new_data_tensor = torch.tensor(last_sequence_scaled, dtype=torch.float32).unsqueeze(0).to(device) # Добавляем batch dimension
    # if 'loaded_model' in locals():
    #     with torch.no_grad(): prediction = loaded_model(new_data_tensor)
    #     print(f"Предсказание для следующего шага: {prediction.item()}")
except FileNotFoundError: print(f"Файл scaler'а {SCALER_SAVE_PATH} не найден.")
except Exception as e: print(f"Ошибка при загрузке scaler'а: {e}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, Subset
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
# >>> НОВЫЕ ИМЕНА для CNN-GRU модели <<<
model_save_name = 'cnn_gru_nn_improved.pth'
scaler_save_name = 'scaler_cnn_gru_improved.pth'

target_col = 'IMO_fwd_log_ret_3d' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ

# Параметры последовательности
SEQUENCE_LENGTH = 25 # Можно попробовать немного увеличить окно

# Параметры разделения и обучения
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 5
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 32
learning_rate = 0.001
epochs = 100 # Макс. кол-во эпох

# >>> ПАРАМЕТРЫ CNN-GRU МОДЕЛИ <<<
CNN_OUT_CHANNELS = 64 # Количество фильтров (признаков) после CNN
CNN_KERNEL_SIZE = 5   # Размер окна свертки (например, 3 или 5)
GRU_HIDDEN_SIZE = 64 # Размер скрытого состояния GRU
GRU_NUM_LAYERS = 1   # Попробуем 1 слой GRU после CNN
MLP_HIDDEN_SIZE = 32 # Размер скрытого слоя в MLP после GRU
DROPOUT_RATE = 0.3   # Dropout

# Параметры улучшений
EARLY_STOPPING_PATIENCE = 50
SCHEDULER_PATIENCE = 10
SCHEDULER_FACTOR = 0.1
WEIGHT_DECAY = 1e-5

random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled = df_cleaned[target_col].values.reshape(-1, 1)
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей ---
def create_sequences(input_data, target_data, sequence_length):
    sequences, targets = [], []
    for i in range(sequence_length, len(input_data)):
        sequences.append(input_data[i-sequence_length:i])
        targets.append(target_data[i])
    return np.array(sequences), np.array(targets)
X_seq, y_seq = create_sequences(X_scaled, y_unscaled, SEQUENCE_LENGTH)
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ ---
print("-" * 30); print(f"Разделение последовательностей на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ...")
X_temp, X_final_test, y_temp, y_final_test, temp_indices, final_test_indices = train_test_split(
    X_seq, y_seq, original_indices_seq, test_size=FINAL_TEST_SET_SIZE, random_state=random_seed, shuffle=True
)
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)


# --- 5. Определение НОВОЙ МОДЕЛИ (CNN-GRU) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate

        # 1D CNN Слой
        # Вход: (N, C_in, L_in) = (batch_size, input_dim, sequence_length)
        # Выход: (N, C_out, L_out) = (batch_size, cnn_out_channels, sequence_length) (с padding='same')
        # Нам нужно сначала поменять местами sequence_length и input_dim
        self.conv1d = nn.Conv1d(in_channels=input_dim,
                                out_channels=cnn_out_channels,
                                kernel_size=kernel_size,
                                padding='same') # 'same' сохраняет длину последовательности
        self.bn1 = nn.BatchNorm1d(cnn_out_channels) # BatchNorm применяется к каналам
        self.relu1 = nn.ReLU()
        self.dropout_cnn = nn.Dropout(dropout_rate)

        # GRU Слой
        # Вход: (N, L, H_in) = (batch_size, sequence_length, cnn_out_channels)
        # Выход hn: (num_layers, N, H_out) = (gru_num_layers, batch_size, gru_hidden_dim)
        # Нам нужно снова поменять местами измерения после CNN
        self.gru = nn.GRU(input_size=cnn_out_channels, # Входной размер GRU = выход CNN
                          hidden_size=gru_hidden_dim,
                          num_layers=gru_num_layers,
                          batch_first=True, # Ожидаем (batch, seq, feature)
                          dropout=dropout_rate if gru_num_layers > 1 else 0)

        # MLP Голова
        self.mlp = nn.Sequential(
            nn.Linear(gru_hidden_dim, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(mlp_hidden, output_dim)
        )

    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_dim)

        # --- CNN Часть ---
        # Меняем местами измерения для Conv1d: (N, C_in, L_in)
        x_cnn = x.permute(0, 2, 1)
        x_cnn = self.conv1d(x_cnn)
        x_cnn = self.bn1(x_cnn)
        x_cnn = self.relu1(x_cnn)
        x_cnn = self.dropout_cnn(x_cnn)
        # Меняем местами обратно для GRU: (N, L_out, C_out)
        x_gru_input = x_cnn.permute(0, 2, 1)

        # --- GRU Часть ---
        # Инициализация скрытого состояния (не обязательно, GRU сделает это по умолчанию)
        # h0 = torch.zeros(self.gru.num_layers, x.size(0), self.gru.hidden_size).to(x.device)
        # out_gru, hn = self.gru(x_gru_input, h0)
        out_gru, hn = self.gru(x_gru_input) # Используем выход последнего шага

        # --- MLP Часть ---
        # Берем выход GRU с последнего временного шага
        # out_gru shape: (batch_size, sequence_length, gru_hidden_dim)
        last_time_step_out = out_gru[:, -1, :]
        # Или можно взять последнее скрытое состояние: hn[-1] для последнего слоя
        # last_time_step_out = hn[-1] # shape: (batch_size, gru_hidden_dim)

        output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2] # Количество признаков

# --- 6. Кросс-валидация ---
kf = KFold(n_splits=K_FOLDS, shuffle=True, random_state=random_seed)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
for fold, (train_idx, val_idx) in enumerate(kf.split(X_temp, y_temp)):
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    # Подготовка данных
    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]
    X_train_fold_tensor = torch.tensor(X_train_fold, dtype=torch.float32).to(device)
    y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold, dtype=torch.float32).to(device)
    y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)
    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True)
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False)

    # >>> Инициализация CNN-GRU модели <<<
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

    # Обучение фолда
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1} (модель CNN-GRU)...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y)
            loss.backward(); optimizer.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0
        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        with torch.no_grad():
            for batch_X_val, batch_y_val in val_fold_loader:
                outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
        avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
        if (epoch + 1) % 10 == 0: print(f'    Фолд {fold+1}, Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        scheduler.step(avg_val_loss)
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено на эпохе {epoch+1} (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель фолда.")
    model.eval()
    with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
    mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
    print(f"  Метрики валидации фолда {fold+1} (лучшая): MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
    fold_mses.append(mse_fold); fold_r2s.append(r2_fold)

# --- Результаты кросс-валидации ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
print(f"Среднее MSE по фолдам: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})")
print(f"Среднее R² по фолдам: {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
print(f"Среднее количество 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")

# --- 7. Обучение ФИНАЛЬНОЙ модели ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU NN + Улучшения) ---")
# Подготовка данных
X_train_final_seq, X_val_final_seq, y_train_final_seq, y_val_final_seq = train_test_split(X_temp, y_temp, test_size=VAL_SPLIT_FOR_FINAL, random_state=random_seed, shuffle=True)
X_train_final_tensor = torch.tensor(X_train_final_seq, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
X_val_final_tensor = torch.tensor(X_val_final_seq, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor); final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True)
final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False)

# >>> Инициализация CNN-GRU финальной модели <<<
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss()
final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
final_scheduler = ReduceLROnPlateau(final_optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

# Обучение финальной модели
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения (модель CNN-GRU)...")
start_time_final = time.time()
for epoch in range(epochs):
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y)
        loss.backward(); final_optimizer.step()
        train_loss_epoch += loss.item(); batches_processed_train += 1
    avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in final_val_loader:
            outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
            val_loss_epoch += loss_val.item(); batches_processed_val += 1
    avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
    if (epoch + 1) % 10 == 0: print(f'    Финал. обуч., Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')
    final_scheduler.step(avg_val_loss)
    if avg_val_loss < best_val_loss_final: best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка фин. обучения!"); break
end_time_final = time.time(); print(f"Финальное обучение завершено на эпохе {epoch+1} (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели ---
if best_model_state_final:
    try:
        torch.save(best_model_state_final, MODEL_SAVE_PATH); print(f"Лучшая финальная модель ({best_epoch_final}) сохранена: {MODEL_SAVE_PATH}")
        final_model.load_state_dict(best_model_state_final)
    except Exception as e: print(f"Ошибка сохранения/загрузки: {e}")
else: print("Предупреждение: Используется последняя модель.")

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ + Сравнение с SMA(5) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
final_model.eval()
# ВАЖНО: X_final_test УЖЕ содержит ПОСЛЕДОВАТЕЛЬНОСТИ масштабированных данных
X_final_test_tensor_eval = torch.tensor(X_final_test, dtype=torch.float32).to(device)
with torch.no_grad(): y_pred_nn = final_model(X_final_test_tensor_eval).cpu().numpy()
# Получение предсказаний SMA(5) для соответствующих дат КОНЦА последовательностей
y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
else: print("NaN в SMA(5) не обнаружено.")

# Расчет метрик
mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
ssr_model = mse_final_nn * len(y_final_test)
mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0

print("\n--- Результаты оценки (CNN-GRU NN + Улучшения) ---")
print(f"MSE NN:               {mse_final_nn:.4f}")
print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}")
print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")
print("---")
print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}")
print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}") # <<< СРАВНЕНИЕ С SMA(5)

# --- 10. Визуализация ---
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
plt.plot([y_final_test.min(), y_final_test.max()], [y_final_test.min(), y_final_test.max()], 'r--', lw=2, label='Идеал')
plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест, CNN-GRU)\nЦель: {target_col}")
plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten(), 'Predicted_SMA5': y_pred_sma5.flatten()}, index=final_test_indices).sort_index()
n_points_to_plot = min(100, len(results_final_test_df))
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)\nЦель: {target_col}")
plt.xticks(rotation=45), plt.legend(), plt.grid(True); plt.tight_layout(); plt.show()
plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (CNN-GRU NN)"); plt.legend(), plt.grid(True); plt.show()

# --- (Дополнительно) Пример загрузки ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
try:
    loaded_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                               gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                               mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE)
    try:
        loaded_model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
        loaded_model.to(device); loaded_model.eval()
        print(f"Модель CNN-GRU успешно загружена из {MODEL_SAVE_PATH}")
    except FileNotFoundError: print(f"Файл модели {MODEL_SAVE_PATH} не найден.")
    except Exception as e: print(f"Ошибка при загрузке модели: {e}")
except NameError: print("Переменная 'input_dim' не определена.")
try:
    with open(SCALER_SAVE_PATH, 'rb') as f: loaded_scaler = pickle.load(f)
    print(f"Scaler успешно загружен из {SCALER_SAVE_PATH}")
    # Пример использования:
    # last_sequence = ... # ваши последние SEQUENCE_LENGTH дней признаков, shape (sequence_length, input_dim)
    # last_sequence_scaled = loaded_scaler.transform(last_sequence)
    # new_data_tensor = torch.tensor(last_sequence_scaled, dtype=torch.float32).unsqueeze(0).to(device) # Добавляем batch dimension
    # if 'loaded_model' in locals():
    #     with torch.no_grad(): prediction = loaded_model(new_data_tensor)
    #     print(f"Предсказание для следующего шага: {prediction.item()}")
except FileNotFoundError: print(f"Файл scaler'а {SCALER_SAVE_PATH} не найден.")
except Exception as e: print(f"Ошибка при загрузке scaler'а: {e}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, Subset
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
# >>> НОВЫЕ ИМЕНА для CNN-GRU модели <<<
model_save_name = 'cnn_gru_nn_improved.pth'
scaler_save_name = 'scaler_cnn_gru_improved.pth'

target_col = 'IMO_fwd_log_ret_3d' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ

# Параметры последовательности
SEQUENCE_LENGTH = 25 # Можно попробовать немного увеличить окно

# Параметры разделения и обучения
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 10
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 32
learning_rate = 0.001
epochs = 3 # Макс. кол-во эпох

# >>> ПАРАМЕТРЫ CNN-GRU МОДЕЛИ <<<
CNN_OUT_CHANNELS = 64 # Количество фильтров (признаков) после CNN
CNN_KERNEL_SIZE = 5   # Размер окна свертки (например, 3 или 5)
GRU_HIDDEN_SIZE = 64 # Размер скрытого состояния GRU
GRU_NUM_LAYERS = 1   # Попробуем 1 слой GRU после CNN
MLP_HIDDEN_SIZE = 32 # Размер скрытого слоя в MLP после GRU
DROPOUT_RATE = 0.3   # Dropout

# Параметры улучшений
EARLY_STOPPING_PATIENCE = 50
SCHEDULER_PATIENCE = 10
SCHEDULER_FACTOR = 0.1
WEIGHT_DECAY = 1e-5

random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled = df_cleaned[target_col].values.reshape(-1, 1)
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей ---
def create_sequences(input_data, target_data, sequence_length):
    sequences, targets = [], []
    for i in range(sequence_length, len(input_data)):
        sequences.append(input_data[i-sequence_length:i])
        targets.append(target_data[i])
    return np.array(sequences), np.array(targets)
X_seq, y_seq = create_sequences(X_scaled, y_unscaled, SEQUENCE_LENGTH)
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ ---
print("-" * 30); print(f"Разделение последовательностей на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ...")
X_temp, X_final_test, y_temp, y_final_test, temp_indices, final_test_indices = train_test_split(
    X_seq, y_seq, original_indices_seq, test_size=FINAL_TEST_SET_SIZE, random_state=random_seed, shuffle=True
)
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)


# --- 5. Определение НОВОЙ МОДЕЛИ (CNN-GRU) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate

        # 1D CNN Слой
        # Вход: (N, C_in, L_in) = (batch_size, input_dim, sequence_length)
        # Выход: (N, C_out, L_out) = (batch_size, cnn_out_channels, sequence_length) (с padding='same')
        # Нам нужно сначала поменять местами sequence_length и input_dim
        self.conv1d = nn.Conv1d(in_channels=input_dim,
                                out_channels=cnn_out_channels,
                                kernel_size=kernel_size,
                                padding='same') # 'same' сохраняет длину последовательности
        self.bn1 = nn.BatchNorm1d(cnn_out_channels) # BatchNorm применяется к каналам
        self.relu1 = nn.ReLU()
        self.dropout_cnn = nn.Dropout(dropout_rate)

        # GRU Слой
        # Вход: (N, L, H_in) = (batch_size, sequence_length, cnn_out_channels)
        # Выход hn: (num_layers, N, H_out) = (gru_num_layers, batch_size, gru_hidden_dim)
        # Нам нужно снова поменять местами измерения после CNN
        self.gru = nn.GRU(input_size=cnn_out_channels, # Входной размер GRU = выход CNN
                          hidden_size=gru_hidden_dim,
                          num_layers=gru_num_layers,
                          batch_first=True, # Ожидаем (batch, seq, feature)
                          dropout=dropout_rate if gru_num_layers > 1 else 0)

        # MLP Голова
        self.mlp = nn.Sequential(
            nn.Linear(gru_hidden_dim, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(mlp_hidden, output_dim)
        )

    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_dim)

        # --- CNN Часть ---
        # Меняем местами измерения для Conv1d: (N, C_in, L_in)
        x_cnn = x.permute(0, 2, 1)
        x_cnn = self.conv1d(x_cnn)
        x_cnn = self.bn1(x_cnn)
        x_cnn = self.relu1(x_cnn)
        x_cnn = self.dropout_cnn(x_cnn)
        # Меняем местами обратно для GRU: (N, L_out, C_out)
        x_gru_input = x_cnn.permute(0, 2, 1)

        # --- GRU Часть ---
        # Инициализация скрытого состояния (не обязательно, GRU сделает это по умолчанию)
        # h0 = torch.zeros(self.gru.num_layers, x.size(0), self.gru.hidden_size).to(x.device)
        # out_gru, hn = self.gru(x_gru_input, h0)
        out_gru, hn = self.gru(x_gru_input) # Используем выход последнего шага

        # --- MLP Часть ---
        # Берем выход GRU с последнего временного шага
        # out_gru shape: (batch_size, sequence_length, gru_hidden_dim)
        last_time_step_out = out_gru[:, -1, :]
        # Или можно взять последнее скрытое состояние: hn[-1] для последнего слоя
        # last_time_step_out = hn[-1] # shape: (batch_size, gru_hidden_dim)

        output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2] # Количество признаков

# --- 6. Кросс-валидация ---
kf = KFold(n_splits=K_FOLDS, shuffle=True, random_state=random_seed)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
for fold, (train_idx, val_idx) in enumerate(kf.split(X_temp, y_temp)):
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    # Подготовка данных
    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]
    X_train_fold_tensor = torch.tensor(X_train_fold, dtype=torch.float32).to(device)
    y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold, dtype=torch.float32).to(device)
    y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)
    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True)
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False)

    # >>> Инициализация CNN-GRU модели <<<
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

    # Обучение фолда
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1} (модель CNN-GRU)...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y)
            loss.backward(); optimizer.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0
        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        with torch.no_grad():
            for batch_X_val, batch_y_val in val_fold_loader:
                outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
        avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
        if (epoch + 1) % 10 == 0: print(f'    Фолд {fold+1}, Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        scheduler.step(avg_val_loss)
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено на эпохе {epoch+1} (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель фолда.")
    model.eval()
    with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
    mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
    print(f"  Метрики валидации фолда {fold+1} (лучшая): MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
    fold_mses.append(mse_fold); fold_r2s.append(r2_fold)

# --- Результаты кросс-валидации ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU NN + Улучшения) ---")
print(f"Среднее MSE по фолдам: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})")
print(f"Среднее R² по фолдам: {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
print(f"Среднее количество 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")

# --- 7. Обучение ФИНАЛЬНОЙ модели ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU NN + Улучшения) ---")
# Подготовка данных
X_train_final_seq, X_val_final_seq, y_train_final_seq, y_val_final_seq = train_test_split(X_temp, y_temp, test_size=VAL_SPLIT_FOR_FINAL, random_state=random_seed, shuffle=True)
X_train_final_tensor = torch.tensor(X_train_final_seq, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
X_val_final_tensor = torch.tensor(X_val_final_seq, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor); final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True)
final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False)

# >>> Инициализация CNN-GRU финальной модели <<<
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                       gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                       mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss()
final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
final_scheduler = ReduceLROnPlateau(final_optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

# Обучение финальной модели
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения (модель CNN-GRU)...")
start_time_final = time.time()
for epoch in range(epochs):
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y)
        loss.backward(); final_optimizer.step()
        train_loss_epoch += loss.item(); batches_processed_train += 1
    avg_train_loss = train_loss_epoch / batches_processed_train if batches_processed_train > 0 else 0; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in final_val_loader:
            outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
            val_loss_epoch += loss_val.item(); batches_processed_val += 1
    avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
    if (epoch + 1) % 10 == 0: print(f'    Финал. обуч., Эпоха [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')
    final_scheduler.step(avg_val_loss)
    if avg_val_loss < best_val_loss_final: best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE: print(f"    Ранняя остановка фин. обучения!"); break
end_time_final = time.time(); print(f"Финальное обучение завершено на эпохе {epoch+1} (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели ---
if best_model_state_final:
    try:
        torch.save(best_model_state_final, MODEL_SAVE_PATH); print(f"Лучшая финальная модель ({best_epoch_final}) сохранена: {MODEL_SAVE_PATH}")
        final_model.load_state_dict(best_model_state_final)
    except Exception as e: print(f"Ошибка сохранения/загрузки: {e}")
else: print("Предупреждение: Используется последняя модель.")

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ + Сравнение с SMA(5) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
final_model.eval()
# ВАЖНО: X_final_test УЖЕ содержит ПОСЛЕДОВАТЕЛЬНОСТИ масштабированных данных
X_final_test_tensor_eval = torch.tensor(X_final_test, dtype=torch.float32).to(device)
with torch.no_grad(): y_pred_nn = final_model(X_final_test_tensor_eval).cpu().numpy()
# Получение предсказаний SMA(5) для соответствующих дат КОНЦА последовательностей
y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
else: print("NaN в SMA(5) не обнаружено.")

# Расчет метрик
mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
ssr_model = mse_final_nn * len(y_final_test)
mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0

print("\n--- Результаты оценки (CNN-GRU NN + Улучшения) ---")
print(f"MSE NN:               {mse_final_nn:.4f}")
print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}")
print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")
print("---")
print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}")
print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}") # <<< СРАВНЕНИЕ С SMA(5)

# --- 10. Визуализация ---
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
plt.plot([y_final_test.min(), y_final_test.max()], [y_final_test.min(), y_final_test.max()], 'r--', lw=2, label='Идеал')
plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест, CNN-GRU)\nЦель: {target_col}")
plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten(), 'Predicted_SMA5': y_pred_sma5.flatten()}, index=final_test_indices).sort_index()
n_points_to_plot = min(100, len(results_final_test_df))
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)\nЦель: {target_col}")
plt.xticks(rotation=45), plt.legend(), plt.grid(True); plt.tight_layout(); plt.show()
plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (CNN-GRU NN)"); plt.legend(), plt.grid(True); plt.show()

# --- (Дополнительно) Пример загрузки ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
try:
    loaded_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE,
                               gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS,
                               mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE)
    try:
        loaded_model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
        loaded_model.to(device); loaded_model.eval()
        print(f"Модель CNN-GRU успешно загружена из {MODEL_SAVE_PATH}")
    except FileNotFoundError: print(f"Файл модели {MODEL_SAVE_PATH} не найден.")
    except Exception as e: print(f"Ошибка при загрузке модели: {e}")
except NameError: print("Переменная 'input_dim' не определена.")
try:
    with open(SCALER_SAVE_PATH, 'rb') as f: loaded_scaler = pickle.load(f)
    print(f"Scaler успешно загружен из {SCALER_SAVE_PATH}")
    # Пример использования:
    # last_sequence = ... # ваши последние SEQUENCE_LENGTH дней признаков, shape (sequence_length, input_dim)
    # last_sequence_scaled = loaded_scaler.transform(last_sequence)
    # new_data_tensor = torch.tensor(last_sequence_scaled, dtype=torch.float32).unsqueeze(0).to(device) # Добавляем batch dimension
    # if 'loaded_model' in locals():
    #     with torch.no_grad(): prediction = loaded_model(new_data_tensor)
    #     print(f"Предсказание для следующего шага: {prediction.item()}")
except FileNotFoundError: print(f"Файл scaler'а {SCALER_SAVE_PATH} не найден.")
except Exception as e: print(f"Ошибка при загрузке scaler'а: {e}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, Subset
from torch.optim.lr_scheduler import ReduceLROnPlateau
# >>> НУЖЕН TimeSeriesSplit <<<
from sklearn.model_selection import train_test_split, KFold, TimeSeriesSplit
# <<<------------------------>>>
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ (Остается как у вас) ---
csv_file_name = 'final_dataset.csv' # <<< Имя файла
model_save_name = 'cnn_gru_nn_improved.pth'
scaler_save_name = 'scaler_cnn_gru_improved.pth'
target_col = 'IMO_fwd_log_ret_3d' # <<< ЦЕЛЕВАЯ ПЕРЕМЕННАЯ
SEQUENCE_LENGTH = 30
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 6
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 32
learning_rate = 0.001
epochs = 3
CNN_OUT_CHANNELS = 64
CNN_KERNEL_SIZE = 5
GRU_HIDDEN_SIZE = 64
GRU_NUM_LAYERS = 1
MLP_HIDDEN_SIZE = 32
DROPOUT_RATE = 0.1
EARLY_STOPPING_PATIENCE =60
SCHEDULER_PATIENCE = 10
SCHEDULER_FACTOR = 0.5
WEIGHT_DECAY = 1e-5
random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}")

# Определение устройства (остается)
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed (остается)
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных (остается) ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
sma5_col_name = f'{target_col}_SMA5'
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец: {sma5_col_name}")

# --- 2. Масштабирование признаков (остается) ---
# ВАЖНО: Scaler обучается на ВСЕХ ДАННЫХ перед разделением.
# Для строгой проверки его нужно обучать ТОЛЬКО на обучающих данных
# внутри фолдов КФ и перед финальным обучением.
# Оставляю ваш вариант, но имейте это в виду для реальных исследований.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cleaned[feature_cols].values)
y_unscaled = df_cleaned[target_col].values.reshape(-1, 1)
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler, f)
    print(f"Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# --- 3. Создание последовательностей (остается) ---
def create_sequences(input_data, target_data, sequence_length):
    sequences, targets = [], []
    for i in range(sequence_length, len(input_data)):
        sequences.append(input_data[i-sequence_length:i])
        targets.append(target_data[i])
    return np.array(sequences), np.array(targets)
X_seq, y_seq = create_sequences(X_scaled, y_unscaled, SEQUENCE_LENGTH)
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")

# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ (ИСПРАВЛЕНО) ---
print("-" * 30); print(f"Разделение данных по времени (без перемешивания)...")
n_samples = X_seq.shape[0]
n_test = int(n_samples * FINAL_TEST_SET_SIZE)
if n_test == 0 and n_samples > 1: n_test = 1 # Минимум 1 для теста
if n_test >= n_samples: raise ValueError("Размер теста равен или больше общего числа образцов.")
n_train_val = n_samples - n_test

# >>> Используем срезы для разделения по времени <<<
X_temp, X_final_test = X_seq[:n_train_val], X_seq[n_train_val:]
y_temp, y_final_test = y_seq[:n_train_val], y_seq[n_train_val:]
temp_indices, final_test_indices = original_indices_seq[:n_train_val], original_indices_seq[n_train_val:]

if X_temp.shape[0] == 0 or X_final_test.shape[0] == 0:
    raise ValueError(f"Ошибка разделения на Temp/Final Test: Temp={X_temp.shape[0]}, Test={X_final_test.shape[0]}")
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)
# --- КОНЕЦ ИСПРАВЛЕНИЯ В ШАГЕ 4 ---

# --- 5. Определение МОДЕЛИ (CNN-GRU) (остается) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate
        self.conv1d = nn.Conv1d(in_channels=input_dim, out_channels=cnn_out_channels, kernel_size=kernel_size, padding='same')
        self.bn1 = nn.BatchNorm1d(cnn_out_channels)
        self.relu1 = nn.ReLU(); self.dropout_cnn = nn.Dropout(dropout_rate)
        self.gru = nn.GRU(input_size=cnn_out_channels, hidden_size=gru_hidden_dim, num_layers=gru_num_layers, batch_first=True, dropout=dropout_rate if gru_num_layers > 1 else 0)
        self.mlp = nn.Sequential(nn.Linear(gru_hidden_dim, mlp_hidden), nn.ReLU(), nn.Dropout(dropout_rate), nn.Linear(mlp_hidden, output_dim))
    def forward(self, x):
        if x.dim() != 3: raise ValueError(f"Expected 3D input (N, L, C), got {x.dim()}")
        x_cnn = x.permute(0, 2, 1); x_cnn = self.conv1d(x_cnn); x_cnn = self.bn1(x_cnn); x_cnn = self.relu1(x_cnn); x_cnn = self.dropout_cnn(x_cnn)
        x_gru_input = x_cnn.permute(0, 2, 1); out_gru, hn = self.gru(x_gru_input)
        last_time_step_out = out_gru[:, -1, :]; output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2]

# --- 6. Кросс-валидация (ИСПРАВЛЕНО) ---
# >>> Используем TimeSeriesSplit <<<
tscv = TimeSeriesSplit(n_splits=K_FOLDS)
fold_mses, fold_r2s = [], []
best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (TimeSeriesSplit) ---")

# ВАЖНО: Масштабирование теперь должно быть ВНУТРИ фолда!
# Мы не можем использовать X_scaled, обученный на всех данных.
# Поэтому оставляем X_temp НЕмасштабированным и масштабируем внутри.
# Это потребует перестройки X_temp в 2D и обратно.

for fold, (train_idx, val_idx) in enumerate(tscv.split(X_temp)): # Сплит на ТЕМП данных
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    print(f"  Train: {train_idx.min()}-{train_idx.max()} ({len(train_idx)}), Val: {val_idx.min()}-{val_idx.max()} ({len(val_idx)})")
    if len(train_idx) == 0 or len(val_idx) == 0: print("  Пропуск фолда."); continue

    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]

    # >>> Масштабирование ВНУТРИ фолда <<<
    n_s_tr, n_l_tr, n_f_tr = X_train_fold.shape
    X_train_fold_2d = X_train_fold.reshape(-1, n_f_tr)
    scaler_fold = StandardScaler()
    scaler_fold.fit(X_train_fold_2d) # Обучаем scaler на трейне фолда
    X_train_fold_scaled_2d = scaler_fold.transform(X_train_fold_2d)
    X_train_fold_scaled = X_train_fold_scaled_2d.reshape(n_s_tr, n_l_tr, n_f_tr)

    n_s_val, n_l_val, n_f_val = X_val_fold.shape
    X_val_fold_2d = X_val_fold.reshape(-1, n_f_val)
    X_val_fold_scaled_2d = scaler_fold.transform(X_val_fold_2d) # Применяем обученный scaler
    X_val_fold_scaled = X_val_fold_scaled_2d.reshape(n_s_val, n_l_val, n_f_val)

    # Создаем тензоры из МАСШТАБИРОВАННЫХ данных фолда
    X_train_fold_tensor = torch.tensor(X_train_fold_scaled, dtype=torch.float32).to(device); y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold_scaled, dtype=torch.float32).to(device); y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)

    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True) # shuffle=True для трейна фолда - ок
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False) if len(val_fold_dataset) > 0 else []

    # Инициализация модели, оптимизатора, шедулера (остается)
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE, gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS, mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss(); optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

    # Обучение фолда с Early Stopping (остается)
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1}...")
    for epoch in range(epochs):
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            if batch_X.dim() != 3: continue
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y); loss.backward(); optimizer.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        if batches_processed_train == 0: continue
        avg_train_loss = train_loss_epoch / batches_processed_train
        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        if len(val_fold_loader) > 0:
            with torch.no_grad():
                for batch_X_val, batch_y_val in val_fold_loader:
                    if batch_X_val.dim() != 3: continue
                    outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                    val_loss_epoch += loss_val.item(); batches_processed_val += 1
            avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
        else: avg_val_loss = float('inf')
        if (epoch + 1) % 50 == 0: print(f'    Эпоха [{epoch+1}/{epochs}], T Loss: {avg_train_loss:.4f}, V Loss: {avg_val_loss:.4f}, LR: {optimizer.param_groups[0]["lr"]:.1e}')
        scheduler.step(avg_val_loss)
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE and len(val_fold_loader) > 0: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда (остается)
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель.");
    model.eval()
    if X_val_fold_tensor.shape[0] > 0:
        with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
        mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
        print(f"  Метрики валидации фолда {fold+1}: MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
        fold_mses.append(mse_fold); fold_r2s.append(r2_fold)
    else: print(f"  Пропуск оценки фолда {fold+1}.")

# --- Результаты кросс-валидации (остается) ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU - TimeSeriesSplit) ---")
if fold_mses:
    print(f"Среднее MSE: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})"); print(f"Среднее R²:  {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
    if best_epochs_per_fold: print(f"Ср. 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")
else: print("Кросс-валидация не дала результатов.")


# --- 7. Обучение ФИНАЛЬНОЙ модели (ИСПРАВЛЕНО) ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU - TimeSeriesSplit) ---")
# >>> Разделение ТЕМП на train_final/val_final ПО ВРЕМЕНИ <<<
n_temp_samples = X_temp.shape[0]
n_val_final = int(n_temp_samples * VAL_SPLIT_FOR_FINAL)
if n_val_final == 0 and n_temp_samples > 1: n_val_final = 1
n_train_final = n_temp_samples - n_val_final
if n_train_final <= 0: raise ValueError("Нет данных для обучения финальной модели.")

X_train_final_seq, X_val_final_seq = X_temp[:n_train_final], X_temp[n_train_final:]
y_train_final_seq, y_val_final_seq = y_temp[:n_train_final], y_temp[n_train_final:]
print(f"  Размер Train Final: {X_train_final_seq.shape[0]}, Val Final: {X_val_final_seq.shape[0]}")

# >>> Обучаем ФИНАЛЬНЫЙ Scaler ТОЛЬКО на train_final <<<
n_s_tr_fin, n_l_tr_fin, n_f_tr_fin = X_train_final_seq.shape
X_train_final_2d = X_train_final_seq.reshape(-1, n_f_tr_fin)
scaler_final = StandardScaler()
scaler_final.fit(X_train_final_2d) # Обучаем финальный scaler
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler_final, f)
    print(f"Финальный Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# Масштабируем train_final и val_final
X_train_final_scaled_2d = scaler_final.transform(X_train_final_2d)
X_train_final_scaled = X_train_final_scaled_2d.reshape(n_s_tr_fin, n_l_tr_fin, n_f_tr_fin)
if X_val_final_seq.shape[0] > 0:
    n_s_val_fin, n_l_val_fin, n_f_val_fin = X_val_final_seq.shape
    X_val_final_2d = X_val_final_seq.reshape(-1, n_f_val_fin)
    X_val_final_scaled_2d = scaler_final.transform(X_val_final_2d)
    X_val_final_scaled = X_val_final_scaled_2d.reshape(n_s_val_fin, n_l_val_fin, n_f_val_fin)
else: X_val_final_scaled = np.array([])

# Тензоры и DataLoader'ы для финального обучения
X_train_final_tensor = torch.tensor(X_train_final_scaled, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor)
if len(final_train_dataset) == 0: raise ValueError("Нет данных для финального обучения.")
final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True)
if X_val_final_scaled.shape[0] > 0:
    X_val_final_tensor = torch.tensor(X_val_final_scaled, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
    final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False)
else: final_val_loader = []

# Инициализация финальной модели, оптимизатора, шедулера (остается)
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE, gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS, mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss(); final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
final_scheduler = ReduceLROnPlateau(final_optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

# Обучение финальной модели с Early Stopping (остается)
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения..."); start_time_final = time.time()
for epoch in range(epochs):
    # ... (код цикла обучения, валидации, early stopping без изменений) ...
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        if batch_X.dim() != 3: continue
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y); loss.backward(); final_optimizer.step() # Использовал optimizer вместо final_optimizer, исправлено
        train_loss_epoch += loss.item(); batches_processed_train += 1
    if batches_processed_train == 0: continue
    avg_train_loss = train_loss_epoch / batches_processed_train; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    if len(final_val_loader) > 0:
        with torch.no_grad():
            for batch_X_val, batch_y_val in final_val_loader:
                if batch_X_val.dim() != 3: continue
                outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
            avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
    else: avg_val_loss = float('inf')
    if (epoch + 1) % 50 == 0: print(f'    Эпоха [{epoch+1}/{epochs}], T Loss: {avg_train_loss:.4f}, V Loss: {avg_val_loss:.4f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}')
    final_scheduler.step(avg_val_loss)
    if avg_val_loss < best_val_loss_final: best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE and len(final_val_loader) > 0: print(f"    Ранняя остановка!"); break
end_time_final = time.time(); print(f"Финальное обучение завершено (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели (остается) ---
if best_model_state_final:
    try: torch.save(best_model_state_final, MODEL_SAVE_PATH); print(f"Лучшая модель ({best_epoch_final}) сохранена: {MODEL_SAVE_PATH}"); final_model.load_state_dict(best_model_state_final)
    except Exception as e: print(f"Ошибка сохранения/загрузки: {e}")
else:
    print("Предупреждение: Используется последняя модель.")
    try: torch.save(final_model.state_dict(), MODEL_SAVE_PATH); print(f"Последняя модель сохранена: {MODEL_SAVE_PATH}")
    except Exception as e: print(f"Ошибка сохранения: {e}")

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ (ИСПРАВЛЕНО масштабирование) ---
# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ ---")
final_model.eval()
if X_final_test.shape[0] > 0:
    # >>> Масштабируем тестовые данные ФИНАЛЬНЫМ scaler'ом <<<
    n_s_test, n_l_test, n_f_test = X_final_test.shape
    X_final_test_2d = X_final_test.reshape(-1, n_f_test)
    # >>> Убедимся, что scaler_final существует и обучен <<<
    if 'scaler_final' not in locals() or not hasattr(scaler_final, 'mean_'):
         print("ОШИБКА: scaler_final не найден или не обучен!")
         # Можно прервать или использовать исходный scaler, если он остался
         # scaler_final = scaler # Использовать старый scaler - ПЛОХО, но для отладки
         raise ValueError("scaler_final не определен!")

    print(f"Финальный тест (до масштабирования): shape={X_final_test_2d.shape}, has NaN={np.isnan(X_final_test_2d).any()}, has inf={np.isinf(X_final_test_2d).any()}")
    X_final_test_scaled_2d = scaler_final.transform(X_final_test_2d)
    print(f"Финальный тест (после масштабирования): shape={X_final_test_scaled_2d.shape}, has NaN={np.isnan(X_final_test_scaled_2d).any()}, has inf={np.isinf(X_final_test_scaled_2d).any()}")

    X_final_test_scaled = X_final_test_scaled_2d.reshape(n_s_test, n_l_test, n_f_test)

    X_final_test_tensor_eval = torch.tensor(X_final_test_scaled, dtype=torch.float32).to(device)
    print(f"Тензор для предсказания: shape={X_final_test_tensor_eval.shape}, device={X_final_test_tensor_eval.device}")

    with torch.no_grad():
        y_pred_nn = final_model(X_final_test_tensor_eval).cpu().numpy()
    print(f"Предсказания NN: shape={y_pred_nn.shape}, has NaN={np.isnan(y_pred_nn).any()}, has inf={np.isinf(y_pred_nn).any()}")
    print(f"Реальные значения y_final_test: shape={y_final_test.shape}, has NaN={np.isnan(y_final_test).any()}, has inf={np.isinf(y_final_test).any()}")


    mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
    print("\n--- Результаты оценки (CNN-GRU) ---")
    print(f"MSE NN:               {mse_final_nn:.4f}"); print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
    print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}"); print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")

    # Сравнение с SMA(5)
    if sma5_col_name in df_cleaned.columns:
        # >>> Проверим индексы <<<
        print(f"Индексы для SMA: len={len(final_test_indices)}, тип={type(final_test_indices)}")
        # print(f"Первые 5 индексов для SMA: {final_test_indices[:5]}")
        # print(f"Последние 5 индексов для SMA: {final_test_indices[-5:]}")
        try:
             y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
             print(f"Предсказания SMA5: shape={y_pred_sma5.shape}, has NaN={np.isnan(y_pred_sma5).any()}")
             mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
             if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
             else: print("NaN в SMA(5) не обнаружено.")

             mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
             ssr_model = mse_final_nn * len(y_final_test)
             r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0
             print("---"); print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}"); print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}")
        except KeyError as e:
             print(f"ОШИБКА: Не удалось получить SMA5 по индексам: {e}")
             print("Возможно, проблема с final_test_indices.")
             y_pred_sma5 = np.full_like(y_final_test, np.nan) # Создаем массив NaN для графика
        except Exception as e:
             print(f"Неизвестная ошибка при работе с SMA5: {e}")
             y_pred_sma5 = np.full_like(y_final_test, np.nan)
    else:
        print("\nСтолбец SMA(5) не найден.")
        y_pred_sma5 = np.full_like(y_final_test, np.nan) # Создаем NaN для графика

    # --- 10. Визуализация ---
    print("\n--- Построение графика результатов ---")
    # >>> Проверим данные прямо перед графиком <<<
    print(f"Данные для графика: y_final_test shape={y_final_test.shape}, y_pred_nn shape={y_pred_nn.shape}, y_pred_sma5 shape={y_pred_sma5.shape}, final_test_indices len={len(final_test_indices)}")

    try:
        plt.figure(figsize=(15, 6))
        # График 1
        plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
        # Добавим проверку на NaN/inf перед расчетом min/max для линии
        valid_limits = np.isfinite(y_final_test)
        if valid_limits.any():
             min_val, max_val = np.nanmin(y_final_test[valid_limits]), np.nanmax(y_final_test[valid_limits])
             plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Идеал')
        else:
             plt.plot([], [], 'r--', lw=2, label='Идеал') # Пустая линия, если нет валидных данных
        plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест)\nЦель: {target_col}, R²={r2_final_nn:.4f}")
        plt.legend(), plt.grid(True);

        # График 2
        plt.subplot(1, 2, 2)
        results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten()}, index=final_test_indices)
        if sma5_col_name in df_cleaned.columns:
             results_final_test_df['Predicted_SMA5'] = y_pred_sma5.flatten()
        results_final_test_df = results_final_test_df.sort_index()
        n_points_to_plot = min(100, len(results_final_test_df))
        if n_points_to_plot > 0:
            plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
            plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
            if sma5_col_name in df_cleaned.columns and not np.isnan(y_pred_sma5).all(): # Рисуем SMA только если она есть и не вся NaN
                 plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
            plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)")
            plt.xticks(rotation=45), plt.legend(), plt.grid(True);
        else: print("Недостаточно точек для графика временного ряда.")

        plt.tight_layout()
        plt.show() # Попробуем снова с show()
        # plot_save_path = os.path.join(base_path, 'cnn_gru_results_plot.png')
        # plt.savefig(plot_save_path, dpi=150, bbox_inches='tight')
        # print(f"График сохранен в {plot_save_path}")
        plt.close()

        # График потерь (если нужен)
        # ...

    except Exception as e:
         print(f"!!! ОШИБКА ПРИ ПОСТРОЕНИИ ГРАФИКА: {e} !!!") # Явное сообщение об ошибке
         # Выведем дополнительную информацию о данных для графика
         print("Доп. инфо о данных для графика:")
         print(f"y_final_test: min={np.nanmin(y_final_test)}, max={np.nanmax(y_final_test)}, has_nan={np.isnan(y_final_test).any()}")
         print(f"y_pred_nn: min={np.nanmin(y_pred_nn)}, max={np.nanmax(y_pred_nn)}, has_nan={np.isnan(y_pred_nn).any()}")
         print(f"y_pred_sma5: min={np.nanmin(y_pred_sma5)}, max={np.nanmax(y_pred_sma5)}, has_nan={np.isnan(y_pred_sma5).any()}")
         import traceback
         traceback.print_exc() # Печатаем полный трейсбэк ошибки графика
else:
    print("Финальный тестовый набор пуст. Оценка и визуализация невозможны.")

# --- (Дополнительно) Пример загрузки (остается) ---
print("\n--- Пример загрузки сохраненной модели и scaler ---")
# ... (код загрузки без изменений) ...
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, Subset
from torch.optim.lr_scheduler import ReduceLROnPlateau
# >>> НУЖЕН TimeSeriesSplit <<<
from sklearn.model_selection import train_test_split, KFold, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib
# matplotlib.use('Agg') # Для Colab обычно не нужно, оставляем интерактивный бэкенд
import matplotlib.pyplot as plt
import os
import time
import copy
import pickle

# --- 0. Настройка среды Google Colab и Конфигурация ---
try:
    from google.colab import drive
    drive.mount('/content/drive')
    base_path = '/content/drive/MyDrive/' # ИЗМЕНИТЕ ПУТЬ, ЕСЛИ НУЖНО
    print(f"Google Drive смонтирован. Базовый путь: {base_path}")
except ModuleNotFoundError:
    print("Google Colab Drive не доступен.")
    base_path = './'

# --- КОНФИГУРАЦИЯ (Ваши значения) ---
csv_file_name = 'final_dataset.csv'
model_save_name = 'cnn_gru_nn_corrected_split.pth' # Новое имя для ясности
scaler_save_name = 'scaler_cnn_gru_corrected_split.pth'
target_col = 'IMO_fwd_log_ret_3d'
SEQUENCE_LENGTH = 25
FINAL_TEST_SET_SIZE = 0.1
K_FOLDS = 5 # Вернул 5, как было в коде с ошибкой, но можно уменьшить для скорости
VAL_SPLIT_FOR_FINAL = 0.1
batch_size = 32
learning_rate = 0.001
epochs = 150 # Макс. кол-во эпох (из вашего кода)
CNN_OUT_CHANNELS = 64
CNN_KERNEL_SIZE = 5
GRU_HIDDEN_SIZE = 64
GRU_NUM_LAYERS = 1
MLP_HIDDEN_SIZE = 32
DROPOUT_RATE = 0.3
EARLY_STOPPING_PATIENCE = 50
SCHEDULER_PATIENCE = 10
SCHEDULER_FACTOR = 0.1
WEIGHT_DECAY = 1e-5
random_seed = 42
# --- КОНЕЦ КОНФИГУРАЦИИ ---

file_path = os.path.join(base_path, csv_file_name)
MODEL_SAVE_PATH = os.path.join(base_path, model_save_name)
SCALER_SAVE_PATH = os.path.join(base_path, scaler_save_name)
print(f"Файл данных: {file_path}"); print(f"Модель: {MODEL_SAVE_PATH}"); print(f"Scaler: {SCALER_SAVE_PATH}")

# Определение устройства
if torch.cuda.is_available(): device = torch.device("cuda"); print(f"GPU: {torch.cuda.get_device_name(0)}")
else: device = torch.device("cpu"); print("CPU.")

# Установка random seed
torch.manual_seed(random_seed); np.random.seed(random_seed)
if device.type == 'cuda': torch.cuda.manual_seed_all(random_seed); torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False

# --- 1. Загрузка и обработка данных ---
try:
    df = pd.read_csv(file_path, index_col='date', parse_dates=True)
    print(f"Данные загружены. Форма: {df.shape}")
except FileNotFoundError: raise SystemExit(f"Ошибка: Файл не найден: {file_path}")
except Exception as e: raise SystemExit(f"Ошибка чтения CSV: {e}")
if target_col not in df.columns: raise ValueError(f"Целевой столбец '{target_col}' не найден!")
feature_cols = [col for col in df.columns if col != target_col]
print(f"Целевая переменная: {target_col}, Признаков: {len(feature_cols)}")
if not feature_cols: raise ValueError("Нет признаков для обучения.")
print(f"Размер данных до удаления NaN: {df.shape}")
df_cleaned = df.dropna(subset=[target_col] + feature_cols).copy()
print(f"Размер данных после удаления NaN: {df_cleaned.shape}")
if df_cleaned.empty: raise ValueError("Нет данных после удаления NaN.")
# Расчет SMA(5) для сравнения (оставляем)
sma5_col_name = f'{target_col}_SMA5_baseline' # Дадим другое имя, чтобы не путать с признаками
df_cleaned[sma5_col_name] = df_cleaned[target_col].rolling(window=5).mean().shift(1)
print(f"Добавлен столбец для сравнения: {sma5_col_name}")

# --- 2. Подготовка X и Y (БЕЗ масштабирования здесь) ---
X_unscaled_2d = df_cleaned[feature_cols].values
y_unscaled_2d = df_cleaned[target_col].values.reshape(-1, 1)

# --- 3. Создание последовательностей ---
def create_sequences(input_data_2d, target_data_2d, sequence_length):
    # (функция без изменений)
    sequences, targets = [], []
    if input_data_2d.ndim != 2 or target_data_2d.ndim != 2: raise ValueError("Input/target must be 2D")
    if input_data_2d.shape[0] != target_data_2d.shape[0]: raise ValueError("Input/target rows mismatch")
    num_samples = input_data_2d.shape[0]
    if num_samples <= sequence_length: return np.array([]), np.array([])
    for i in range(sequence_length, num_samples):
        sequences.append(input_data_2d[i-sequence_length:i, :])
        targets.append(target_data_2d[i, :])
    return (np.array(sequences), np.array(targets)) if sequences else (np.array([]), np.array([]))

X_seq, y_seq = create_sequences(X_unscaled_2d, y_unscaled_2d, SEQUENCE_LENGTH)
if X_seq.shape[0] == 0: raise ValueError("Не удалось создать последовательности.")
original_indices_seq = df_cleaned.index[SEQUENCE_LENGTH:]
print(f"Созданы последовательности: X_seq shape={X_seq.shape}, y_seq shape={y_seq.shape}")


# --- 4. Разделение данных на ТЕМП и ФИНАЛЬНЫЙ ТЕСТ (ИСПРАВЛЕНО: ПО ВРЕМЕНИ) ---
print("-" * 30); print(f"Разделение данных по времени (без перемешивания)...")
n_samples = X_seq.shape[0]
n_test = int(n_samples * FINAL_TEST_SET_SIZE)
if n_test == 0 and n_samples > 1: n_test = 1
if n_test >= n_samples: raise ValueError("Размер теста >= общего числа образцов.")
n_train_val = n_samples - n_test

# >>> Используем срезы для разделения <<<
X_temp, X_final_test = X_seq[:n_train_val], X_seq[n_train_val:]
y_temp, y_final_test = y_seq[:n_train_val], y_seq[n_train_val:]
temp_indices, final_test_indices = original_indices_seq[:n_train_val], original_indices_seq[n_train_val:]

if X_temp.shape[0] == 0 or X_final_test.shape[0] == 0: raise ValueError("Ошибка разделения на Temp/Final Test.")
print(f"Размер ТЕМП X: {X_temp.shape}, ФИНАЛЬНЫЙ ТЕСТ X: {X_final_test.shape}"); print("-" * 30)
# --- КОНЕЦ ИСПРАВЛЕНИЯ В ШАГЕ 4 ---


# --- 5. Определение МОДЕЛИ (CNN-GRU) (остается) ---
class CNNGRUModel(nn.Module):
    def __init__(self, input_dim, cnn_out_channels, kernel_size, gru_hidden_dim, gru_num_layers, mlp_hidden, output_dim=1, dropout_rate=0.0):
        super(CNNGRUModel, self).__init__()
        self.dropout_rate = dropout_rate
        self.conv1d = nn.Conv1d(in_channels=input_dim, out_channels=cnn_out_channels, kernel_size=kernel_size, padding='same')
        self.bn1 = nn.BatchNorm1d(cnn_out_channels); self.relu1 = nn.ReLU(); self.dropout_cnn = nn.Dropout(dropout_rate)
        self.gru = nn.GRU(input_size=cnn_out_channels, hidden_size=gru_hidden_dim, num_layers=gru_num_layers, batch_first=True, dropout=dropout_rate if gru_num_layers > 1 else 0)
        self.mlp = nn.Sequential(nn.Linear(gru_hidden_dim, mlp_hidden), nn.ReLU(), nn.Dropout(dropout_rate), nn.Linear(mlp_hidden, output_dim))
    def forward(self, x):
        if x.dim() != 3: raise ValueError(f"Expected 3D input (N, L, C), got {x.dim()}")
        x_cnn = x.permute(0, 2, 1); x_cnn = self.conv1d(x_cnn); x_cnn = self.bn1(x_cnn); x_cnn = self.relu1(x_cnn); x_cnn = self.dropout_cnn(x_cnn)
        x_gru_input = x_cnn.permute(0, 2, 1); out_gru, hn = self.gru(x_gru_input)
        last_time_step_out = out_gru[:, -1, :]; output = self.mlp(last_time_step_out)
        return output

input_dim = X_seq.shape[2]

# --- 6. Кросс-валидация (ИСПРАВЛЕНО: TimeSeriesSplit + Масштабирование внутри) ---
# >>> Используем TimeSeriesSplit <<<
tscv = TimeSeriesSplit(n_splits=K_FOLDS)
fold_mses, fold_r2s = [], []; best_epochs_per_fold = []
print(f"\n--- Начало {K_FOLDS}-кратной кросс-валидации (TimeSeriesSplit) ---")

for fold, (train_idx, val_idx) in enumerate(tscv.split(X_temp)): # Сплит на ТЕМП данных
    print(f"\n--- ФОЛД {fold+1}/{K_FOLDS} ---")
    print(f"  Train: {train_idx.min()}-{train_idx.max()} ({len(train_idx)}), Val: {val_idx.min()}-{val_idx.max()} ({len(val_idx)})")
    if len(train_idx) == 0 or len(val_idx) == 0: print("  Пропуск фолда."); continue

    X_train_fold, X_val_fold = X_temp[train_idx], X_temp[val_idx]
    y_train_fold, y_val_fold = y_temp[train_idx], y_temp[val_idx]

    # >>> Масштабирование ВНУТРИ фолда <<<
    n_s_tr, n_l_tr, n_f_tr = X_train_fold.shape
    X_train_fold_2d = X_train_fold.reshape(-1, n_f_tr)
    scaler_fold = StandardScaler()
    scaler_fold.fit(X_train_fold_2d) # Обучаем scaler на трейне фолда
    X_train_fold_scaled_2d = scaler_fold.transform(X_train_fold_2d)
    X_train_fold_scaled = X_train_fold_scaled_2d.reshape(n_s_tr, n_l_tr, n_f_tr)

    n_s_val, n_l_val, n_f_val = X_val_fold.shape
    X_val_fold_2d = X_val_fold.reshape(-1, n_f_val)
    X_val_fold_scaled_2d = scaler_fold.transform(X_val_fold_2d) # Применяем обученный scaler
    X_val_fold_scaled = X_val_fold_scaled_2d.reshape(n_s_val, n_l_val, n_f_val)

    # Создаем тензоры из МАСШТАБИРОВАННЫХ данных фолда
    X_train_fold_tensor = torch.tensor(X_train_fold_scaled, dtype=torch.float32).to(device); y_train_fold_tensor = torch.tensor(y_train_fold, dtype=torch.float32).to(device)
    X_val_fold_tensor = torch.tensor(X_val_fold_scaled, dtype=torch.float32).to(device); y_val_fold_tensor = torch.tensor(y_val_fold, dtype=torch.float32).to(device)

    train_fold_dataset = TensorDataset(X_train_fold_tensor, y_train_fold_tensor); train_fold_loader = DataLoader(train_fold_dataset, batch_size=batch_size, shuffle=True) # shuffle=True для трейна фолда - ок
    val_fold_dataset = TensorDataset(X_val_fold_tensor, y_val_fold_tensor); val_fold_loader = DataLoader(val_fold_dataset, batch_size=batch_size, shuffle=False) if len(val_fold_dataset) > 0 else []

    # Инициализация модели, оптимизатора, шедулера (остается)
    model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE, gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS, mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
    criterion = nn.MSELoss(); optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

    # Обучение фолда с Early Stopping (остается)
    best_val_loss = float('inf'); epochs_no_improve = 0; best_model_state = None; best_epoch = -1
    start_time_fold = time.time(); print(f"  Начало обучения фолда {fold+1}...")
    for epoch in range(epochs):
        # ... (код цикла обучения и валидации без изменений) ...
        model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
        for batch_X, batch_y in train_fold_loader:
            if batch_X.dim() != 3: continue
            optimizer.zero_grad(); outputs = model(batch_X); loss = criterion(outputs, batch_y); loss.backward(); optimizer.step()
            train_loss_epoch += loss.item(); batches_processed_train += 1
        if batches_processed_train == 0: continue
        avg_train_loss = train_loss_epoch / batches_processed_train
        model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
        if len(val_fold_loader) > 0:
            with torch.no_grad():
                for batch_X_val, batch_y_val in val_fold_loader:
                    if batch_X_val.dim() != 3: continue
                    outputs_val = model(batch_X_val); loss_val = criterion(outputs_val, batch_y_val)
                    val_loss_epoch += loss_val.item(); batches_processed_val += 1
            avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
        else: avg_val_loss = float('inf')
        if (epoch + 1) % 50 == 0: print(f'    Эпоха [{epoch+1}/{epochs}], T Loss: {avg_train_loss:.4f}, V Loss: {avg_val_loss:.4f}, LR: {optimizer.param_groups[0]["lr"]:.1e}') # Увеличил шаг вывода
        scheduler.step(avg_val_loss)
        if avg_val_loss < best_val_loss: best_val_loss = avg_val_loss; best_model_state = copy.deepcopy(model.state_dict()); best_epoch = epoch + 1; epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= EARLY_STOPPING_PATIENCE and len(val_fold_loader) > 0: print(f"    Ранняя остановка!"); break
    end_time_fold = time.time(); print(f"  Обучение фолда {fold+1} завершено (лучший {best_epoch}) за {end_time_fold - start_time_fold:.2f} сек.")
    best_epochs_per_fold.append(best_epoch)

    # Оценка лучшей модели фолда (остается)
    if best_model_state: model.load_state_dict(best_model_state)
    else: print("  Предупреждение: Используется последняя модель.");
    model.eval()
    if X_val_fold_tensor.shape[0] > 0:
        with torch.no_grad(): y_val_pred = model(X_val_fold_tensor).cpu().numpy()
        mse_fold = mean_squared_error(y_val_fold, y_val_pred); r2_fold = r2_score(y_val_fold, y_val_pred)
        print(f"  Метрики валидации фолда {fold+1}: MSE={mse_fold:.4f}, R²={r2_fold:.4f}")
        fold_mses.append(mse_fold); fold_r2s.append(r2_fold)
    else: print(f"  Пропуск оценки фолда {fold+1}.")

# --- Результаты кросс-валидации (остается) ---
print("\n--- Сводные результаты К-кратной кросс-валидации (CNN-GRU - TimeSeriesSplit) ---")
if fold_mses:
    print(f"Среднее MSE: {np.mean(fold_mses):.4f} (+/- {np.std(fold_mses):.4f})"); print(f"Среднее R²:  {np.mean(fold_r2s):.4f} (+/- {np.std(fold_r2s):.4f})")
    if best_epochs_per_fold: print(f"Ср. 'лучших' эпох: {np.mean(best_epochs_per_fold):.1f} (+/- {np.std(best_epochs_per_fold):.1f})")
else: print("Кросс-валидация не дала результатов.")


# --- 7. Обучение ФИНАЛЬНОЙ модели (ИСПРАВЛЕНО: Разделение и Масштабирование) ---
print("\n--- Обучение финальной модели на ТЕМП данных (CNN-GRU - TimeSeriesSplit) ---")
# >>> Разделение ТЕМП на train_final/val_final ПО ВРЕМЕНИ <<<
n_temp_samples = X_temp.shape[0]
# Расчет доли валидации от n_temp_samples
n_val_final = int(n_temp_samples * VAL_SPLIT_FOR_FINAL)
if n_val_final == 0 and n_temp_samples > 1: n_val_final = 1 # Минимум 1 для валидации
n_train_final = n_temp_samples - n_val_final
if n_train_final <= 0: raise ValueError("Нет данных для обучения финальной модели.")

X_train_final_seq, X_val_final_seq = X_temp[:n_train_final], X_temp[n_train_final:]
y_train_final_seq, y_val_final_seq = y_temp[:n_train_final], y_temp[n_train_final:]
print(f"  Размер Train Final: {X_train_final_seq.shape[0]}, Val Final: {X_val_final_seq.shape[0]}")

# >>> Обучаем ФИНАЛЬНЫЙ Scaler ТОЛЬКО на train_final <<<
n_s_tr_fin, n_l_tr_fin, n_f_tr_fin = X_train_final_seq.shape
X_train_final_2d = X_train_final_seq.reshape(-1, n_f_tr_fin)
scaler_final = StandardScaler()
scaler_final.fit(X_train_final_2d) # Обучаем финальный scaler
try:
    with open(SCALER_SAVE_PATH, 'wb') as f: pickle.dump(scaler_final, f)
    print(f"Финальный Scaler сохранен: {SCALER_SAVE_PATH}")
except Exception as e: print(f"Ошибка сохранения scaler: {e}")

# Масштабируем train_final и val_final
X_train_final_scaled_2d = scaler_final.transform(X_train_final_2d)
X_train_final_scaled = X_train_final_scaled_2d.reshape(n_s_tr_fin, n_l_tr_fin, n_f_tr_fin)
if X_val_final_seq.shape[0] > 0:
    n_s_val_fin, n_l_val_fin, n_f_val_fin = X_val_final_seq.shape
    X_val_final_2d = X_val_final_seq.reshape(-1, n_f_val_fin)
    X_val_final_scaled_2d = scaler_final.transform(X_val_final_2d)
    X_val_final_scaled = X_val_final_scaled_2d.reshape(n_s_val_fin, n_l_val_fin, n_f_val_fin)
else: X_val_final_scaled = np.array([])

# Тензоры и DataLoader'ы для финального обучения
X_train_final_tensor = torch.tensor(X_train_final_scaled, dtype=torch.float32).to(device); y_train_final_tensor = torch.tensor(y_train_final_seq, dtype=torch.float32).to(device)
final_train_dataset = TensorDataset(X_train_final_tensor, y_train_final_tensor)
if len(final_train_dataset) == 0: raise ValueError("Нет данных для финального обучения.")
final_train_loader = DataLoader(final_train_dataset, batch_size=batch_size, shuffle=True) # shuffle=True для трейна финальной модели - ок

if X_val_final_scaled.shape[0] > 0:
    X_val_final_tensor = torch.tensor(X_val_final_scaled, dtype=torch.float32).to(device); y_val_final_tensor = torch.tensor(y_val_final_seq, dtype=torch.float32).to(device)
    final_val_dataset = TensorDataset(X_val_final_tensor, y_val_final_tensor); final_val_loader = DataLoader(final_val_dataset, batch_size=batch_size, shuffle=False)
else: final_val_loader = []

# Инициализация финальной модели, оптимизатора, шедулера (остается)
final_model = CNNGRUModel(input_dim=input_dim, cnn_out_channels=CNN_OUT_CHANNELS, kernel_size=CNN_KERNEL_SIZE, gru_hidden_dim=GRU_HIDDEN_SIZE, gru_num_layers=GRU_NUM_LAYERS, mlp_hidden=MLP_HIDDEN_SIZE, dropout_rate=DROPOUT_RATE).to(device)
final_criterion = nn.MSELoss(); final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=WEIGHT_DECAY)
final_scheduler = ReduceLROnPlateau(final_optimizer, mode='min', factor=SCHEDULER_FACTOR, patience=SCHEDULER_PATIENCE, verbose=False)

# Обучение финальной модели с Early Stopping (остается)
best_val_loss_final = float('inf'); epochs_no_improve_final = 0; best_model_state_final = None; best_epoch_final = -1; final_train_losses_history = []
print(f"  Начало финального обучения..."); start_time_final = time.time()
for epoch in range(epochs):
    # ... (код цикла обучения и валидации без изменений) ...
    final_model.train(); train_loss_epoch = 0.0; batches_processed_train = 0
    for batch_X, batch_y in final_train_loader:
        if batch_X.dim() != 3: continue
        final_optimizer.zero_grad(); outputs = final_model(batch_X); loss = final_criterion(outputs, batch_y); loss.backward(); final_optimizer.step() # Исправлено на final_optimizer
        train_loss_epoch += loss.item(); batches_processed_train += 1
    if batches_processed_train == 0: continue
    avg_train_loss = train_loss_epoch / batches_processed_train; final_train_losses_history.append(avg_train_loss)
    final_model.eval(); val_loss_epoch = 0.0; batches_processed_val = 0
    if len(final_val_loader) > 0:
        with torch.no_grad():
            for batch_X_val, batch_y_val in final_val_loader:
                if batch_X_val.dim() != 3: continue
                outputs_val = final_model(batch_X_val); loss_val = final_criterion(outputs_val, batch_y_val)
                val_loss_epoch += loss_val.item(); batches_processed_val += 1
            avg_val_loss = val_loss_epoch / batches_processed_val if batches_processed_val > 0 else float('inf')
    else: avg_val_loss = float('inf')
    if (epoch + 1) % 50 == 0: print(f'    Эпоха [{epoch+1}/{epochs}], T Loss: {avg_train_loss:.4f}, V Loss: {avg_val_loss:.4f}, LR: {final_optimizer.param_groups[0]["lr"]:.1e}') # Увеличил шаг вывода
    final_scheduler.step(avg_val_loss)
    if avg_val_loss < best_val_loss_final: best_val_loss_final = avg_val_loss; best_model_state_final = copy.deepcopy(final_model.state_dict()); best_epoch_final = epoch + 1; epochs_no_improve_final = 0
    else:
        epochs_no_improve_final += 1
        if epochs_no_improve_final >= EARLY_STOPPING_PATIENCE and len(final_val_loader) > 0: print(f"    Ранняя остановка!"); break
end_time_final = time.time(); print(f"Финальное обучение завершено (лучший {best_epoch_final}) за {end_time_final - start_time_final:.2f} сек.")

# --- 8. Сохранение ЛУЧШЕЙ ФИНАЛЬНОЙ модели (остается) ---
if best_model_state_final:
    try: torch.save(best_model_state_final, MODEL_SAVE_PATH); print(f"Лучшая модель ({best_epoch_final}) сохранена: {MODEL_SAVE_PATH}"); final_model.load_state_dict(best_model_state_final)
    except Exception as e: print(f"Ошибка сохранения/загрузки: {e}")
else:
    print("Предупреждение: Используется последняя модель.")
    try: torch.save(final_model.state_dict(), MODEL_SAVE_PATH); print(f"Последняя модель сохранена: {MODEL_SAVE_PATH}")
    except Exception as e: print(f"Ошибка сохранения: {e}")

# --- 9. Оценка ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ТЕСТЕ (ИСПРАВЛЕНО масштабирование) ---
print("\n--- Оценка ЛУЧШЕЙ ФИНАЛЬНОЙ модели на ОТДЕЛЬНОМ ФИНАЛЬНОМ ТЕСТЕ (CNN-GRU) ---")
final_model.eval()
if X_final_test.shape[0] > 0:
    # >>> Масштабируем тестовые данные ФИНАЛЬНЫМ scaler'ом <<<
    n_s_test, n_l_test, n_f_test = X_final_test.shape
    X_final_test_2d = X_final_test.reshape(-1, n_f_test)
    # >>> Используем scaler_final <<<
    X_final_test_scaled_2d = scaler_final.transform(X_final_test_2d)
    X_final_test_scaled = X_final_test_scaled_2d.reshape(n_s_test, n_l_test, n_f_test)

    X_final_test_tensor_eval = torch.tensor(X_final_test_scaled, dtype=torch.float32).to(device)
    with torch.no_grad(): y_pred_nn = final_model(X_final_test_tensor_eval).cpu().numpy()
    mse_final_nn = mean_squared_error(y_final_test, y_pred_nn); r2_final_nn = r2_score(y_final_test, y_pred_nn)
    print("\n--- Результаты оценки (CNN-GRU) ---")
    print(f"MSE NN:               {mse_final_nn:.4f}"); print(f"RMSE NN:              {np.sqrt(mse_final_nn):.4f}")
    print(f"MAE NN:               {mean_absolute_error(y_final_test, y_pred_nn):.4f}"); print(f"R² NN (vs Mean):      {r2_final_nn:.4f}")

    # Сравнение с SMA(5) (остается)
    if sma5_col_name in df_cleaned.columns:
        try: # Добавим try-except на случай проблем с индексами
            y_pred_sma5 = df_cleaned.loc[final_test_indices, sma5_col_name].values.reshape(-1, 1)
            mean_y_test = np.mean(y_final_test); nan_mask_sma5 = np.isnan(y_pred_sma5)
            if np.any(nan_mask_sma5): print(f"Заполнение {np.sum(nan_mask_sma5)} NaN в SMA(5)..."); y_pred_sma5[nan_mask_sma5] = mean_y_test
            else: print("NaN в SMA(5) не обнаружено.")
            mse_final_sma5 = mean_squared_error(y_final_test, y_pred_sma5); ssr_baseline_sma5 = mse_final_sma5 * len(y_final_test)
            ssr_model = mse_final_nn * len(y_final_test)
            r2_vs_sma5 = 1 - (ssr_model / ssr_baseline_sma5) if ssr_baseline_sma5 > 1e-9 else 0.0
            print("---"); print(f"MSE Базовой SMA(5):   {mse_final_sma5:.4f}"); print(f"R² NN (vs SMA(5)):    {r2_vs_sma5:.4f}")
        except KeyError: print(f"Не удалось найти индексы для SMA5 в df_cleaned.") ; y_pred_sma5 = np.full_like(y_final_test, np.nan)
        except Exception as e: print(f"Ошибка при расчете SMA5: {e}"); y_pred_sma5 = np.full_like(y_final_test, np.nan)
    else: print("\nСтолбец SMA(5) не найден."); y_pred_sma5 = np.full_like(y_final_test, np.nan)


    # --- 10. Визуализация (остается, но с проверкой y_pred_sma5) ---
    print("\n--- Построение графика результатов ---")
    try:
        plt.figure(figsize=(15, 6))
        plt.subplot(1, 2, 1); plt.scatter(y_final_test, y_pred_nn, alpha=0.6, label='Предсказание CNN-GRU vs Реальность')
        valid_limits = np.isfinite(y_final_test); min_val, max_val = (np.nanmin(y_final_test[valid_limits]), np.nanmax(y_final_test[valid_limits])) if valid_limits.any() else (0,0)
        plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Идеал')
        plt.xlabel("Реальные"), plt.ylabel("Предсказанные (CNN-GRU)"); plt.title(f"Сравнение (Фин. тест)\nЦель: {target_col}, R²={r2_final_nn:.4f}")
        plt.legend(), plt.grid(True); plt.subplot(1, 2, 2)
        results_final_test_df = pd.DataFrame({'Real': y_final_test.flatten(), 'Predicted_NN': y_pred_nn.flatten()}, index=final_test_indices)
        if sma5_col_name in df_cleaned.columns and not np.isnan(y_pred_sma5).all(): results_final_test_df['Predicted_SMA5'] = y_pred_sma5.flatten()
        results_final_test_df = results_final_test_df.sort_index()
        n_points_to_plot = min(100, len(results_final_test_df))
        if n_points_to_plot > 0:
            plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Real'][:n_points_to_plot], 'b-o', label='Реальные', markersize=4, alpha=0.7)
            plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_NN'][:n_points_to_plot], 'r-x', label='Предсказания CNN-GRU', markersize=4, alpha=0.7)
            if 'Predicted_SMA5' in results_final_test_df.columns: plt.plot(results_final_test_df.index[:n_points_to_plot], results_final_test_df['Predicted_SMA5'][:n_points_to_plot], 'g--^', label='Предсказания SMA(5)', markersize=4, alpha=0.7)
            plt.xlabel("Дата"), plt.ylabel("Значение " + target_col); plt.title(f"Предсказания vs Реальность ({n_points_to_plot} точек теста)")
            plt.xticks(rotation=45), plt.legend(), plt.grid(True);
        plt.tight_layout(); plt.show()
        plt.close() # Закрываем фигуру результатов

        if 'final_train_losses_history' in locals() and final_train_losses_history:
            plt.figure(figsize=(10, 5)); plt.plot(range(1, len(final_train_losses_history) + 1), final_train_losses_history, label='MSE Обучения')
            plt.xlabel("Эпоха"), plt.ylabel("Средняя ошибка"); plt.title("Динамика ошибки финального обучения (CNN-GRU)"); plt.legend(), plt.grid(True); plt.show()
            plt.close() # Закрываем фигуру потерь
    except Exception as e: print(f"Ошибка визуализации: {e}")
else: print("Финальный тест пуст.")

# --- (Дополнительно) Пример загрузки (остается) ---
# ... (код загрузки без изменений) ...
```
